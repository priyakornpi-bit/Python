## Main Event Loop (ลูปเหตุการณ์หลัก)

ใน PySide6 การสร้าง GUI จะต้องเริ่มจากการสร้างอินสแตนซ์ของ `QApplication` ก่อนที่จะสร้างหน้าต่างใดๆ จากนั้นจึงเรียก `app.exec()` เพื่อเริ่มลูปเหตุการณ์หลัก

### เหตุผลที่ต้องมี QApplication:
- **QApplication จัดการการตั้งค่าพื้นฐาน**: จัดการกับการตั้งค่าพื้นฐานของแอปพลิเคชัน เช่น การจัดการหน่วยความจำและทรัพยากร
- **จัดการเหตุการณ์**: จัดการกับการส่งต่อเหตุการณ์ (events) ไปยัง Widget ต่างๆ เช่น การคลิกเมาส์หรือการกดแป้นพิมพ์
- **รองรับการตั้งค่า**: รองรับการตั้งค่าพื้นฐานเช่น ภาษาและธีมของระบบ

### การทำงานของ Event Loop:
- **รอรับเหตุการณ์**: Event Loop จะรอรับเหตุการณ์ต่างๆ ที่เกิดขึ้นในแอปพลิเคชัน
- **ส่งต่อเหตุการณ์**: เมื่อมีเหตุการณ์เกิดขึ้น จะส่งต่อไปยัง Widget ที่เกี่ยวข้องเพื่อให้ทำงานตามที่กำหนด
- **ทำงานต่อเนื่อง**: กระบวนการนี้จะทำงานต่อเนื่องจนกว่าผู้ใช้จะปิดแอปพลิเคชันหรือเรียก `app.quit()`

### ตัวอย่างโค้ดพื้นฐาน:
```python
from PySide6.QtWidgets import QApplication
import sys

app = QApplication(sys.argv)  # สร้าง QApplication
# สร้างหน้าต่างและ Widget ที่นี่
window.show()  # แสดงหน้าต่าง
app.exec()  # เริ่ม Event Loop
```

### คำเตือน:
- QApplication ต้องมีเพียงหนึ่งเดียวต่อแอปพลิเคชัน
- `sys.argv` ใช้สำหรับส่งพารามิเตอร์จาก command line หากไม่ต้องการสามารถใช้ `[]` แทนได้

````python
# Import required modules for PySide6
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget, QLabel, QPushButton, QVBoxLayout, QHBoxLayout, QGridLayout, QFormLayout, QLineEdit, QTextEdit, QSpinBox, QComboBox, QCheckBox, QMessageBox
from PySide6.QtCore import Qt
from PySide6.QtGui import QFont
import sys
````

````python
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget
import sys

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Main Event Loop Example")
        self.setGeometry(100, 100, 400, 300)
        
        central_widget = QWidget()
        self.setCentralWidget(central_widget)

app = QApplication([])
window = MainWindow()
window.show()
app.exec()
````

## วิดเจ็ต (Widgets)

วิดเจ็ตคือองค์ประกอบภาพใดๆ ในแอปพลิเคชัน Qt ที่ผู้ใช้สามารถเห็นหรือโต้ตอบด้วย เช่น ปุ่มกล่องข้อความ หรือป้ายกำกับ วิดเจ็ตเป็นส่วนประกอบพื้นฐานในการสร้างอินเทอร์เฟซผู้ใช้

### ลักษณะของวิดเจ็ต:
- **ส่วนแสดงผล (Appearance)**: จัดการรูปลักษณ์ เช่น สี ขนาด ตำแหน่ง และการแสดงผล
- **ส่วนเหตุการณ์ (Event)**: จัดการการโต้ตอบ เช่น การคลิกเมาส์ การพิมพ์ข้อความ
- **การจัดกลุ่ม**: สามารถมีวิดเจ็ตอื่นๆ เป็นลูกได้ ทำให้สามารถสร้างโครงสร้างที่ซับซ้อน
- **การสืบทอด**: PySide ใช้การสืบทอดคลาสอย่างเข้มงวด ทำให้โค้ดมีโครงสร้างที่ดี

### ประเภทของวิดเจ็ต:
- **วิดเจ็ตพื้นฐาน**: QLabel (ป้ายกำกับ), QPushButton (ปุ่ม), QLineEdit (กล่องข้อความบรรทัดเดียว)
- **วิดเจ็ตคอนเทนเนอร์**: QWidget (คอนเทนเนอร์ทั่วไป), QGroupBox (กล่องกลุ่ม)
- **วิดเจ็ตพิเศษ**: QComboBox (กล่องเลือกแบบดรอปดาวน์), QSpinBox (กล่องตัวเลข), QTextEdit (กล่องข้อความหลายบรรทัด)

### การใช้งานวิดเจ็ต:
```python
button = QPushButton("คลิกฉัน", parent)
button.clicked.connect(callback)  # เชื่อมต่อสัญญาณกับฟังก์ชัน
```

### สัญญาณและสล็อต (Signals and Slots):
- **สัญญาณ (Signal)**: เหตุการณ์ที่วิดเจ็ตส่งออกเมื่อมีอะไรเกิดขึ้น เช่น `clicked` เมื่อคลิกปุ่ม
- **สล็อต (Slot)**: ฟังก์ชันหรือเมธอดที่ทำงานเมื่อได้รับสัญญาณ
- ใช้เมธอด `connect()` เพื่อเชื่อมต่อสัญญาณกับสล็อต ทำให้การโต้ตอบเป็นไปอย่างราบรื่น

### ตัวอย่างการใช้งาน:
```python
from PySide6.QtWidgets import QPushButton, QLabel

# สร้างปุ่มและป้ายกำกับ
button = QPushButton("คลิกที่นี่")
label = QLabel("ข้อความ")

# เชื่อมต่อสัญญาณ
button.clicked.connect(lambda: label.setText("ปุ่มถูกคลิกแล้ว!"))
```

### วิดเจ็ตพื้นฐาน

- **การกำหนด Parent (Widget Parenting)**: ใน PySide การกำหนด parent ให้กับวิดเจ็ตจะทำอย่างชัดเจน ซึ่งช่วยให้การจัดการหน่วยความจำเป็นไปอย่างมีประสิทธิภาพและป้องกันการรั่วไหลของหน่วยความจำ
- **การสืบทอดคลาส**: PySide ให้ความสำคัญกับการสืบทอดคลาส ทำให้โค้ดมีโครงสร้างที่ดี บำรุงรักษาง่าย และสอดคล้องกับหลักการ OOP
- **วิดเจ็ตเฉพาะทาง**: PySide มีวิดเจ็ตเฉพาะทางมากมาย ทำให้ง่ายต่อการสร้าง UI ที่มีรูปลักษณ์มาตรฐานและเป็นมืออาชีพ
- **สัญญาณ/สล็อต**: ใช้ระบบสัญญาณและสล็อตในการจัดการเหตุการณ์ของผู้ใช้ ทำให้การเขียนโค้ดเป็นระเบียบและแยกส่วนได้ดี

### ประโยชน์ของวิดเจ็ตพื้นฐาน:
- **ลดเวลาในการพัฒนา**: ไม่ต้องสร้างวิดเจ็ตตั้งแต่ต้น
- **รูปลักษณ์สอดคล้อง**: มีลักษณะที่เป็นมาตรฐานตามระบบปฏิบัติการ
- **ง่ายต่อการปรับแต่ง**: สามารถปรับแต่งได้ตามความต้องการ
- **รองรับการเข้าถึง**: รองรับการใช้งานสำหรับผู้ใช้พิการ

### ตัวอย่างวิดเจ็ตพื้นฐาน:
- **QPushButton**: ปุ่มสำหรับคลิกเพื่อดำเนินการ
- **QLabel**: ป้ายกำกับสำหรับแสดงข้อความหรือรูปภาพ
- **QLineEdit**: กล่องข้อความบรรทัดเดียวสำหรับป้อนข้อมูล

- **QTextEdit**: กล่องข้อความหลายบรรทัดสำหรับแก้ไขข้อความยาว

### การจัดการเหตุการณ์:
```python
button = QPushButton("คลิก")
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget, QPushButton, QVBoxLayout
import sys

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Signals and Slots Example")
        self.setGeometry(100, 100, 300, 200)
        
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        
        layout = QVBoxLayout(central_widget)

        layout.addWidget(self.button)

    def handle_click(self):

```

### เคล็ดลับการใช้งาน:
- ควรตั้งชื่อวิดเจ็ตให้สื่อความหมายเพื่อความง่ายในการบำรุงรักษา
- ใช้ layout เพื่อจัดวางวิดเจ็ตให้เป็นระเบียบ
- ทดสอบการทำงานของวิดเจ็ตในทุกสถานการณ์ที่เป็นไปได้

````python
button = QPushButton("Click", parent)
button.clicked.connect(callback) # signal and slot
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget, QLabel, QLineEdit, QVBoxLayout
import sys

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Widget Configuration Example")
        self.setGeometry(100, 100, 400, 200)
        
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        
        layout = QVBoxLayout(central_widget)
        
        # สร้าง widget พื้นฐาน
        label = QLabel("Hello World")
        label.setFont(QFont("Arial", 14, QFont.Bold))
        label.setAlignment(Qt.AlignCenter)
        
        entry = QLineEdit()
        entry.setPlaceholderText("Enter text here...")
        entry.setMaxLength(50)
        
        layout.addWidget(label)
        layout.addWidget(entry)

# ทดสอบ (สำหรับ notebook เท่านั้น)
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    app.exec()
````

### การกำหนดค่าวิดเจ็ต

- **การใช้ Setter Methods**: PySide ใช้เมธอด setter ที่ชัดเจนในการกำหนดค่า เช่น `button.setText("ข้อความ")` แทนการกำหนดค่าโดยตรง
- **การพิมพ์แบบเข้มงวด**: คุณสมบัติของ PySide มีการพิมพ์แบบเข้มงวด ทำให้มั่นใจได้ว่าข้อมูลที่กำหนดมีชนิดถูกต้อง
- **การตรวจสอบข้อผิดพลาด**: หากพยายามกำหนดค่าที่ไม่ถูกต้อง เช่น `button.setEnabled("false")` แทน `False` หรือ `button.setEnabled(0)` PySide จะเกิด TypeError ซึ่งช่วยให้ตรวจพบข้อผิดพลาดได้ตั้งแต่เนิ่นๆ และมั่นใจว่าคุณสมบัติถูกกำหนดด้วยชนิดข้อมูลที่ถูกต้อง

### ตัวอย่างการกำหนดค่า:
```python
button.setText("คลิกฉัน")        # กำหนดข้อความบนปุ่ม
button.setEnabled(False)         # ปิดใช้งานปุ่ม
button.setFixedSize(100, 30)     # กำหนดขนาดคงที่
button.setStyleSheet("color: red;")  # กำหนดสไตล์ CSS
```

### ประโยชน์ของการพิมพ์แบบเข้มงวด:
- **ลดข้อผิดพลาด**: ช่วยตรวจพบข้อผิดพลาดในการพัฒนาได้ตั้งแต่เนิ่นๆ
- **โค้ดอ่านง่าย**: ทำให้โค้ดเข้าใจง่ายและบำรุงรักษาง่าย
- **IDE Support**: ช่วยให้ IDE สามารถให้คำแนะนำและตรวจสอบได้ดีขึ้น
- **ความน่าเชื่อถือ**: มั่นใจในความถูกต้องของข้อมูลที่ใช้

## การจัดการ Layout

- **Layout Objects แยกต่างหาก**: PySide ใช้วัตถุ layout แยกต่างหากจาก widget ซึ่งมีประสิทธิภาพและยืดหยุ่นมากกว่าการจัดการตำแหน่งด้วยตนเอง ทำให้ง่ายต่อการปรับเปลี่ยนและบำรุงรักษา
- **Layout Types ที่หลากหลาย**: PySide มี layout ประเภทต่างๆ มากมายเพื่อรองรับการจัดวางที่แตกต่างกัน
- **การควบคุม Spacing และ Margins**: การจัดการระยะห่างระหว่าง widget และขอบเขตทำได้ง่ายและละเอียดใน PySide

### ประเภทของ Layout:
- **QFormLayout**: เหมาะสำหรับฟอร์มที่มี label และ field จับคู่กัน เช่น ฟอร์มลงทะเบียน
- **QStackedLayout**: ใช้สำหรับสลับแสดง widget ต่างๆ ในพื้นที่เดียวกัน เช่น wizard หรือ tab
- **QVBoxLayout**: จัดวาง widget แนวตั้ง (vertical) จากบนลงล่าง
- **QHBoxLayout**: จัดวาง widget แนวนอน (horizontal) จากซ้ายไปขวา
- **QGridLayout**: จัดวาง widget ในรูปแบบตาราง (grid) เหมาะสำหรับการจัดวางที่ซับซ้อน

### ตัวอย่างการใช้งาน:
```python
layout = QVBoxLayout(parent_widget)  # สร้าง layout แนวตั้ง
layout.addWidget(label)              # เพิ่ม widget ป้ายกำกับ
layout.addWidget(button)             # เพิ่ม widget ปุ่ม
layout.addWidget(text_edit)          # เพิ่ม widget กล่องข้อความ
```

### การจัดการ Spacing และ Margins:
- `setSpacing(int)`: กำหนดระยะห่างระหว่าง widget ใน layout
- `setContentsMargins(left, top, right, bottom)`: กำหนดขอบเขตภายใน layout
- `addStretch(int)`: เพิ่มพื้นที่ยืดหยุ่นเพื่อกระจาย widget
- `addSpacing(int)`: เพิ่มระยะห่างคงที่ระหว่าง widget

### ประโยชน์ของการใช้ Layout:
- **การปรับขนาดอัตโนมัติ**: จัดวาง widget อัตโนมัติเมื่อหน้าต่างเปลี่ยนขนาด
- **UI ที่ตอบสนอง**: ทำให้อินเทอร์เฟซปรับตัวตามการเปลี่ยนแปลงขนาดหน้าต่าง
- **ลดความซับซ้อน**: ไม่ต้องคำนวณตำแหน่งด้วยตนเอง
- **การออกแบบ Responsive**: รองรับการออกแบบที่ปรับตัวตามอุปกรณ์ต่างๆ

### เคล็ดลับการใช้งาน Layout:
- เลือกประเภท layout ที่เหมาะสมกับโครงสร้าง UI ที่ต้องการ
- ใช้ nested layout (layout ซ้อน layout) สำหรับโครงสร้างที่ซับซ้อน
- ทดสอบการปรับขนาดหน้าต่างเพื่อดูการตอบสนองและปรับแต่ง spacing
- ใช้ `addStretch()` เพื่อกระจายพื้นที่ว่างอย่างเหมาะสม
- ตั้งชื่อ layout ให้สื่อความหมายเพื่อความง่ายในการบำรุงรักษา

````python
layout = QVBoxLayout(parent_widget) # create a layout
layout.addWidget(label) # add widget to the layout
layout.addWidget(button) # add more widget to the layout

# Or grid layout in PySide
grid_layout = QGridLayout(parent_widget) # create another layout
grid_layout.addWidget(label, 0, 0) # add widget to the layout
grid_layout.addWidget(button, 1, 0) # add more widget to the layout
````

## หน้าต่าง (Windows)

- **Central Widget**: หน้าต่างหลักของ PySide (QMainWindow) จะต้องมี central widget ซึ่งเป็นพื้นที่หลักสำหรับแสดงเนื้อหาและ widget อื่นๆ
- **Window Types**: PySide มีหน้าต่างเฉพาะทางหลายประเภทเพื่อรองรับการใช้งานที่แตกต่างกัน

### ประเภทของหน้าต่าง:
- **QMainWindow**: หน้าต่างหลักที่มีโครงสร้างครบถ้วน รวมถึงเมนู แถบเครื่องมือ แถบสถานะ และพื้นที่ทำงานกลาง เหมาะสำหรับแอปพลิเคชันเดสก์ท็อปทั่วไป
- **QWidget as window**: หน้าต่างทั่วไปที่สร้างจาก QWidget สามารถปรับแต่งได้อย่างอิสระ เหมาะสำหรับหน้าต่างย่อยหรือหน้าต่างพิเศษ
- **QDialog**: หน้าต่างโต้ตอบ เช่น กล่องโต้ตอบสำหรับการยืนยัน การตั้งค่า หรือการป้อนข้อมูล มีลักษณะ modal (บล็อกหน้าต่างอื่น) หรือ non-modal

### การสร้างหน้าต่างหลัก:
```python
class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("ชื่อแอปพลิเคชัน")  # ตั้งชื่อหน้าต่าง
        self.setGeometry(100, 100, 800, 600)  # ตำแหน่งและขนาด
        
        # สร้าง central widget
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        
        # เพิ่ม layout และ widget อื่นๆ ใน central_widget
```

### ลักษณะและฟีเจอร์ของหน้าต่าง:
- **Title Bar**: แสดงชื่อและมีปุ่มควบคุม (ปิด ย่อ ขยาย)
- **Border และ Frame**: ขอบเขตและกรอบของหน้าต่าง
- **การปรับขนาด**: สามารถปรับขนาดและย้ายได้ตามต้องการ
- **การจัดการสถานะ**: จัดการกับการปิด/ซ่อน/แสดงหน้าต่าง
- **Modal/Non-modal**: หน้าต่าง modal จะบล็อกการโต้ตอบกับหน้าต่างอื่น

### หน้าต่างย่อย (Sub-windows):
```python
class SubWindow(QWidget):
    def __init__(self, parent=None):
        super().__init__(parent)
        # QWidget สามารถทำหน้าที่เป็นหน้าต่างย่อยได้
```

### เคล็ดลับการใช้งาน:
- ใช้ QMainWindow สำหรับหน้าต่างหลักของแอปพลิเคชัน
- ใช้ QDialog สำหรับหน้าต่างโต้ตอบหรือการตั้งค่า
- ตั้งชื่อหน้าต่าง (`setWindowTitle()`) และไอคอน (`setWindowIcon()`) ให้เหมาะสม
- จัดการกับการปิดหน้าต่างโดยใช้ `closeEvent()` หากต้องการ
- พิจารณาใช้ modal dialogs สำหรับการยืนยันที่สำคัญ

## QApplication

มาเริ่มสร้าง GUI โดยใช้ PySide6 กันเถอะ!

QApplication เป็นคลาสหลักที่จัดการแอปพลิเคชัน Qt โดยรวม มีหน้าที่สำคัญดังนี้:
- จัดการ Event Loop ซึ่งเป็นหัวใจของการทำงานของ GUI
- จัดการการตั้งค่าระบบและการปรับแต่งต่างๆ
- จัดการทรัพยากรของแอปพลิเคชัน เช่น หน่วยความจำและการเชื่อมต่อ
- รองรับการจัดการหลายหน้าต่างและการสื่อสารระหว่าง widget

### ขั้นตอนการสร้างแอปพลิเคชัน Qt พื้นฐาน:
1. **Import คลาสที่ต้องการ**: นำเข้า QApplication และ widget ต่างๆ ที่จะใช้
2. **สร้าง QApplication instance**: สร้างออบเจ็กต์ QApplication ซึ่งต้องมีเพียงหนึ่งเดียวต่อแอปพลิเคชัน
3. **สร้างและตั้งค่าหน้าต่างหลัก**: สร้างหน้าต่างหลักและเพิ่ม widget ต่างๆ
4. **แสดงหน้าต่าง**: เรียก `show()` เพื่อแสดงหน้าต่าง
5. **เริ่ม Event Loop**: เรียก `app.exec()` เพื่อเริ่มลูปเหตุการณ์

### ตัวอย่างโค้ดพื้นฐาน:
```python
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget
import sys

# สร้าง QApplication
app = QApplication(sys.argv)

# สร้างหน้าต่างหลัก
window = QMainWindow()
window.setWindowTitle("แอปพลิเคชันแรกของฉัน")
window.setGeometry(100, 100, 400, 300)  # x, y, width, height

# สร้าง central widget
central_widget = QWidget()
window.setCentralWidget(central_widget)

# แสดงหน้าต่าง
window.show()

# เริ่ม event loop
app.exec()
```

### คำเตือนสำคัญ:
- QApplication ต้องมีเพียงหนึ่งเดียวต่อแอปพลิเคชัน หากสร้างหลายตัวอาจเกิดปัญหา
- ต้องเรียก `app.exec()` เพื่อเริ่ม event loop มิฉะนั้นแอปจะไม่ตอบสนองต่อการโต้ตอบของผู้ใช้
- การออกจากแอปพลิเคชันจะเกิดขึ้นเมื่อ event loop สิ้นสุด (โดยปกติเมื่อปิดหน้าต่างหลัก)

### การตั้งค่าเพิ่มเติมสำหรับ QApplication:
- `setStyle(style_name)`: ตั้งสไตล์ของแอปพลิเคชัน (เช่น "Fusion", "Windows")
- `setPalette(palette)`: ตั้งชุดสีของแอปพลิเคชัน
- `setFont(font)`: ตั้งฟอนต์เริ่มต้นสำหรับ widget ทุกตัว
- `setApplicationName(name)`: ตั้งชื่อแอปพลิเคชัน
- `setOrganizationName(name)`: ตั้งชื่อองค์กรสำหรับการตั้งค่า

### เคล็ดลับ:
- ใช้ `sys.argv` เพื่อรองรับ argument จาก command line
- จัดการ exception ใน event loop เพื่อป้องกันแอปพลิเคชัน crash
- ใช้ `app.aboutToQuit` signal เพื่อทำความสะอาดก่อนปิดแอป

## ขั้นตอนการสร้างแอปพลิเคชัน Qt

### 1. Import โมดูลที่จำเป็น
```
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget
import sys
```
- `QApplication`: จัดการแอปพลิเคชันโดยรวม
- `QMainWindow`: หน้าต่างหลักที่มีโครงสร้างครบถ้วน
- `QWidget`: วิดเจ็ตพื้นฐาน
- `sys`: สำหรับการจัดการ argument จาก command line

```
app = QApplication(sys.argv)
```

### 3. สร้างหน้าต่างหลัก
```
window = QMainWindow()
```
- `QMainWindow` มีโครงสร้างสำหรับแอปพลิเคชันเดสก์ท็อปครบถ้วน
- รวมถึงพื้นที่สำหรับเมนู, แถบเครื่องมือ, แถบสถานะ, และพื้นที่ทำงานกลาง

### 4. สร้างและตั้งค่า Central Widget
```
central_widget = QWidget()
window.setCentralWidget(central_widget)
```
- **Central Widget**: เป็นพื้นที่เนื้อหาหลักของหน้าต่าง
- เป็นคอนเทนเนอร์หลักสำหรับ widget อื่นๆ และ layout
- ทุกอย่างที่ต้องการแสดงในหน้าต่างหลักจะถูกเพิ่มใน central widget

### 5. แสดงหน้าต่างหลัก
```
window.show()
```
- ทำให้หน้าต่างปรากฏบนหน้าจอ
- หน้าต่างจะถูกซ่อนโดยค่าเริ่มต้น จนกว่าจะเรียก `show()`

### 6. เริ่ม Event Loop
```
app.exec()
```
- เริ่มลูปเหตุการณ์หลักของแอปพลิเคชัน
- แอปพลิเคชันจะรอรับและจัดการเหตุการณ์ต่างๆ เช่น การคลิกเมาส์, การกดแป้นพิมพ์
- `app.exec()` จะทำงานต่อเนื่องจนกว่าผู้ใช้จะปิดแอปพลิเคชัน

### วิธีที่ 1: Flat Style (สไตล์แบน)

- **เหมาะสำหรับ**: UI ง่ายๆ หรือการทดสอบ
- **จุดเด่น**: แสดงขั้นตอนหลักของการทำงานกับ Qt components ได้ชัดเจน
- **ข้อจำกัด**: ไม่เหมาะสำหรับแอปพลิเคชันที่ซับซ้อน

#### จุดเด่นของ Flat Style:
- **โค้ดสั้นและตรงไปตรงมา**: เขียนทุกอย่างในฟังก์ชันเดียว
- **เข้าใจง่าย**: เหมาะสำหรับผู้เริ่มต้นเรียนรู้ PySide6
- **พัฒนาเร็ว**: ไม่ต้องออกแบบโครงสร้างคลาส
- **แสดงขั้นตอนชัด**: เห็นการทำงานของแต่ละ component ได้อย่างชัดเจน

#### จุดด้อยของ Flat Style:
- **ไม่เป็นระเบียบ**: เมื่อโค้ดยาวขึ้นจะอ่านยาก
- **ยากต่อการบำรุงรักษา**: ยากต่อการแก้ไขหรือเพิ่มฟีเจอร์
- **ไม่สามารถ reuse**: ไม่สามารถนำโค้ดกลับมาใช้ใหม่ได้
- **ผสม logic กับ UI**: จัดการทั้ง logic และ UI ในที่เดียวกัน

#### ตัวอย่าง Flat Style:

- **การเรียนรู้**: เมื่อเริ่มเรียนรู้ PySide6 เป็นครั้งแรก
- **Prototype**: การสร้าง prototype หรือ MVP อย่างรวดเร็ว
- **แอปพลิเคชันขนาดเล็ก**: โปรแกรมเล็กๆ ที่ไม่ซับซ้อน
- **การทดสอบ**: ทดสอบฟีเจอร์หรือ widget ใหม่ๆ
- **Script**: โปรแกรมที่รันครั้งเดียวหรือไม่ต้องการบำรุงรักษา

#### เมื่อไม่ควรใช้ Flat Style:

### วิธีที่ 2: Conventional Class-based Style (สไตล์อิงคลาสแบบเดิม)

**แนะนำให้ใช้วิธีนี้เสมอเมื่อเป็นไปได้**

#### จุดเด่นของ Class-based Style:
- **เป็นระเบียบ**: จัดวางโค้ดอย่างเป็นระบบและอ่านง่าย
- **บำรุงรักษาง่าย**: ง่ายต่อการแก้ไข เพิ่มเติม หรือปรับปรุง
- **Reuse ได้**: สามารถนำคลาสกลับมาใช้ใหม่ในโปรเจ็กต์อื่น
- **ตรงตามหลัก OOP**: ใช้หลักการเขียนโปรแกรมเชิงวัตถุ
- **แยกส่วนได้ดี**: แยก UI, logic, และการจัดการแอปพลิเคชัน

#### โครงสร้างการจัดวางโค้ด:

   - สร้างแอปพลิเคชัน
   - เริ่ม event loop และจัดการ exit code

2. **คลาส Application (QApplication)**:
   - สร้างหน้าต่างหลัก
   - แสดงหน้าต่างหลัก
   - จัดการการตั้งค่าของแอปพลิเคชัน

3. **คลาส Main Window (QMainWindow)**:
   - สร้างและจัดการ widget ต่างๆ
   - จัดการ layout และการจัดวาง
   - เมธอดสำหรับจัดการเหตุการณ์และการโต้ตอบ

# วิดเจ็ต (Widgets)

หลังจากที่เราได้เรียนรู้โครงสร้างพื้นฐานของแอปพลิเคชัน Qt แล้ว ตอนนี้ถึงเวลาสำรวจ **วิดเจ็ต (Widgets)** ซึ่งเป็นองค์ประกอบหลักในการสร้างอินเทอร์เฟซผู้ใช้

วิดเจ็ตคือองค์ประกอบภาพใน GUI ที่ผู้ใช้สามารถเห็นและโต้ตอบด้วย เช่น ปุ่ม, ช่องข้อความ, รายการเลือก เป็นต้น วิดเจ็ตเป็นส่วนสำคัญที่ทำให้แอปพลิเคชันสามารถสื่อสารกับผู้ใช้ได้

## ลักษณะสำคัญของวิดเจ็ต:

### 1. **ส่วนแสดงผล (Appearance)**:
- กำหนดรูปลักษณ์ เช่น ขนาด สี ฟอนต์
- จัดการการแสดงผลบนหน้าจอ

### 2. **ส่วนเหตุการณ์ (Event Handling)**:
- จัดการการโต้ตอบจากผู้ใช้ เช่น การคลิก การพิมพ์
- เชื่อมต่อกับสัญญาณและสล็อต

### 3. **การจัดวาง (Layout)**:
- วิดเจ็ตสามารถมีวิดเจ็ตอื่นเป็นลูกได้
- ใช้ layout เพื่อจัดวางวิดเจ็ตอย่างเป็นระเบียบ

## ประเภทของวิดเจ็ต:

### 1. **วิดเจ็ตแสดงข้อมูล**:
- `QLabel`: แสดงข้อความหรือรูปภาพ
- `QProgressBar`: แสดงความคืบหน้า
- `QLCDNumber`: แสดงตัวเลขแบบ LCD

### 2. **วิดเจ็ตรับข้อมูล**:
- `QLineEdit`: ช่องข้อความบรรทัดเดียว
- `QTextEdit`: ช่องข้อความหลายบรรทัด
- `QSpinBox`: กล่องตัวเลขพร้อมปุ่มเพิ่ม/ลด
- `QDoubleSpinBox`: กล่องตัวเลขทศนิยม

### 3. **วิดเจ็ตเลือกตัวเลือก**:
- `QCheckBox`: ช่องทำเครื่องหมาย
- `QRadioButton`: ปุ่มเลือกเดียวในกลุ่ม
- `QComboBox`: กล่องเลือกแบบดรอปดาวน์
- `QListWidget`: รายการที่เลือกได้

### 4. **วิดเจ็ตดำเนินการ**:
- `QPushButton`: ปุ่มสำหรับคลิก
- `QToolButton`: ปุ่มเครื่องมือ
- `QCommandLinkButton`: ปุ่มลิงก์คำสั่ง

### 5. **วิดเจ็ตจัดกลุ่มและคอนเทนเนอร์**:
- `QGroupBox`: กล่องกลุ่ม
- `QTabWidget`: แท็บหลายหน้า
- `QScrollArea`: พื้นที่เลื่อน

### 6. **วิดเจ็ตแสดงรายการและตาราง**:
- `QTableWidget`: ตารางข้อมูล
- `QTreeWidget`: ต้นไม้ข้อมูล

## การใช้งานวิดเจ็ตทั่วไป:

### ขั้นตอนการใช้งาน:
1. **Import วิดเจ็ต**: จาก `PySide6.QtWidgets`
2. **สร้าง instance**: `widget = QWidgetType()`
3. **ตั้งค่า properties**: `widget.setProperty(value)`
4. **เพิ่มใน layout**: `layout.addWidget(widget)`
5. **เชื่อมต่อสัญญาณ**: `widget.signal.connect(slot)` (ถ้าจำเป็น)

### ตัวอย่างการใช้งาน QLabel:
```python
from PySide6.QtWidgets import QLabel, QVBoxLayout

# สร้าง label
label = QLabel("สวัสดี PySide6!")

# ตั้งค่าคุณสมบัติ
label.setAlignment(Qt.AlignCenter)
label.setStyleSheet("color: blue; font-size: 14px;")

# เพิ่มใน layout
layout.addWidget(label)
```

## สัญญาณและสล็อต (Signals and Slots):

ระบบสัญญาณและสล็อตเป็นวิธีการจัดการเหตุการณ์ใน PySide6:
- **สัญญาณ (Signal)**: เหตุการณ์ที่วิดเจ็ตส่งออก เช่น `clicked`, `textChanged`
- **สล็อต (Slot)**: ฟังก์ชันที่ทำงานเมื่อได้รับสัญญาณ
- **การเชื่อมต่อ**: ใช้ `connect()` เพื่อเชื่อมสัญญาณกับสล็อต

### ตัวอย่าง:
```python
def on_button_clicked():
    print("ปุ่มถูกคลิก!")

button.clicked.connect(on_button_clicked)
```

## เคล็ดลับการใช้งานวิดเจ็ต:

- **ตั้งชื่อให้สื่อความหมาย**: `self.name_label` แทน `self.label1`
- **ใช้ layout เสมอ**: หลีกเลี่ยงการกำหนดตำแหน่งด้วยตนเอง
- **ตั้งค่าเริ่มต้น**: กำหนดค่าที่เหมาะสมตั้งแต่แรก
- **จัดการหน่วยความจำ**: PySide จัดการให้อัตโนมัติเมื่อมี parent
- **ทดสอบการโต้ตอบ**: ตรวจสอบว่า widget ตอบสนองถูกต้อง

ในหัวข้อต่อไป เราจะเรียนรู้เกี่ยวกับ QLabel ซึ่งเป็นวิดเจ็ตพื้นฐานสำหรับแสดงข้อความและรูปภาพ 📝

A widget is any **visual element** in a Qt application that the user **can see or interact** with.

Widgets in PySide has a "Appearance" part and an "Event" part.

**Appearance**:
- Can be configured (color, size, position) - a bit easy and flexible to config
- Can contain other widgets
- Usually put in a layout type

**Event**:
- Can detect mouse, keyboard and other events

You need to import corresponding widget classes from PySide6.QtWidgets

**Common widgets:**
- QWidget
- QLabel (Label)
- QPushButton (Button)
- QLineEdit (Entry)
- QCheckBox (Checkbox)
- QComboBox (Combobox, dropdown)
- QSpinBox (Number with up and down buttons)
- QMessageBox (Pop up message)
- QTextEdit (Multi-line entry)

##QLabel

For text and image **display**

**Text label**
```
text_label = QLabel("Hello World")
text_label.setFont(QFont("Arial", 14, QFont.Bold))
text_label.setAlignment(Qt.AlignCenter)
```

**Image label**
```
image_label = QLabel()
pixmap = QPixmap("image.png")
image_label.setPixmap(pixmap.scaled(
    200, 200,  # width, height
    Qt.KeepAspectRatio,  # maintain aspect ratio
    Qt.SmoothTransformation  # smooth scaling
))
```

**Rich text label**
```
rich_label = QLabel()
rich_label.setText("<h1>Title</h1><p>This is <b>bold</b> and <i>italic</i> text</p>")
rich_label.setTextFormat(Qt.RichText)
rich_label.setOpenExternalLinks(True)  # Allow clicking links
```

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow, QLabel,
                             QVBoxLayout, QWidget)
from PySide6.QtCore import Qt
from PySide6.QtGui import QPixmap, QFont

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PySide6 Demo App")
        self.setGeometry(100, 100, 400, 500)

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        # ----------------------------------------------------
        # Text label
        # ----------------------------------------------------
        # 1) Create Widget
        text_label = QLabel("Hello World")
        text_label.setFont(QFont("Arial", 14, QFont.Bold))
        text_label.setAlignment(Qt.AlignCenter)

        # 2) Add the widget to layout
        layout.addWidget(text_label)

        # ----------------------------------------------------
        # Image label
        # ----------------------------------------------------
        # 1) Create Widget
        image_label = QLabel()
        try:
            pixmap = QPixmap("qrcode.jpg")
            image_label.setPixmap(pixmap.scaled(
                200, 200,  # width, height
                Qt.KeepAspectRatio,  # maintain aspect ratio
                Qt.SmoothTransformation  # smooth scaling
            ))
        except:
            # Fallback if image not found
            image_label.setText("Image not found: image.png")
            image_label.setAlignment(Qt.AlignCenter)

        image_label.setAlignment(Qt.AlignCenter)

        # 2) Add the widget to layout
        layout.addWidget(image_label)

        # ----------------------------------------------------
        # Rich text label
        # ----------------------------------------------------
        # 1) Create Widget
        rich_label = QLabel()
        rich_label.setText("<h1>Title</h1><p>This is <b>bold</b> and <i>italic</i> text</p>")
        rich_label.setTextFormat(Qt.RichText)
        rich_label.setOpenExternalLinks(True)  # Allow clicking links
        rich_label.setAlignment(Qt.AlignCenter)

        # 2) Add the widget to layout
        layout.addWidget(rich_label)

def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

## QPushButton

Interactive Button

**Basic setup**

```
self.setText("Click Me") # text on button
self.setIcon(QIcon("icon.png")) # icon on button
self.setCheckable(True)  # Toggle button
```

**Size configurations**
```
self.setFixedSize(120, 40)  # Fixed width and height
```

**Connect signals**

We will learn about this later. I have included these such that you can see that they are clickable.

```
self.clicked.connect(self.handle_click)

def handle_click(self):
  print("Button clicked!")

def handle_toggle(self, checked):
```

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow,
                             QVBoxLayout, QWidget)
from PySide6.QtWidgets import QPushButton
from PySide6.QtGui import QIcon
from PySide6.QtCore import Qt, QSize

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PySide6 Demo App")
        self.setGeometry(100, 100, 400, 500)

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        # 1) Create an instance of the button class
        button = CustomButton()

        # 2) Add button to layout
        layout.addWidget(button,alignment=Qt.AlignCenter)


class CustomButton(QPushButton):
    def __init__(self):
        super().__init__()

        # Basic setup
        self.setText("Click Me")
        self.setIcon(QIcon("icon.png"))
        self.setCheckable(True)  # Toggle button

        # Size configurations
        self.setFixedSize(120, 40)  # Fixed width and height
        self.setIconSize(QSize(20, 20))

        # Connect signals
        self.clicked.connect(self.handle_click)
        self.toggled.connect(self.handle_toggle)

    def handle_click(self):
        print("Button clicked!")

    def handle_toggle(self, checked):
        print(f"Button toggled: {checked}")

def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

##QLineEdit

Single Line Text Input

**Basic**

```
self.setPlaceholderText("Enter text here...")

```

**Input validation**

Ex. only allow letters and numbers


**Connect signals**
```

def on_text_changed(self, text):
```

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow,
                             QVBoxLayout, QWidget)
from PySide6.QtWidgets import QLineEdit
from PySide6.QtGui import QRegularExpressionValidator

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PySide6 Demo App")
        self.setGeometry(100, 100, 400, 500)

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        # 1) Create an instance of LineEdit
        entry = CustomLineEdit()

        # 2) Add LineEdit to layout
        layout.addWidget(entry)

class CustomLineEdit(QLineEdit):
    def __init__(self):
        super().__init__()

        # Basic setup
        self.setPlaceholderText("Enter text here...")
        self.setMaxLength(50)  # Maximum characters

        # Input validation
        # Only allow letters and numbers
        regex = QRegularExpressionValidator(r"[A-Za-z0-9]+")
        self.setValidator(regex)

        # Echo modes
        self.setEchoMode(QLineEdit.Password)  # For passwords

        # Connect signals
        self.textChanged.connect(self.on_text_changed)
        self.returnPressed.connect(self.on_return_pressed)

    def on_text_changed(self, text):
        print(f"Text changed: {text}")

    def on_return_pressed(self):
        print("Enter key pressed")

def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

##QCheckBox

- Checkbox(es)
- Toggleable Option
- Provide multiple selections

**Basic setup**

```
self.setText("Enable feature")
self.setTristate(True)  # Allow partial checked state
```

**Connect signals**

```
self.stateChanged.connect(self.on_state_changed)

def on_state_changed(self, state):
  states = {
      0: "Unchecked",
      1: "Partially checked",
      2: "Checked"
  }
  print(f"Checkbox state: {states.get(state)}")
```

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow, QWidget, QVBoxLayout,
                               QHBoxLayout, QCheckBox, QPushButton, QGroupBox,
                               QMessageBox)
from PySide6.QtCore import Qt

class CheckboxOptionsWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Checkbox Options Example")
        self.setMinimumSize(400, 250)

        # Main widget and layout
        central_widget = QWidget()
        main_layout = QVBoxLayout(central_widget)

        # 1) Create options group (all checkboxes)
        self.create_food_preferences_group()

        # 2) Button to get selected options
        submit_button = QPushButton("Submit Choices")
        submit_button.clicked.connect(self.show_selected_options)

        # 3) Add widgets to main layout
        main_layout.addWidget(self.food_group)
        main_layout.addWidget(submit_button)
        main_layout.addStretch() # space spreading management

        self.setCentralWidget(central_widget)

    def create_food_preferences_group(self):
        self.food_group = QGroupBox("Food Preferences")
        layout = QVBoxLayout()

        # Create multiple checkboxes
        self.pizza_cb = QCheckBox("Pizza")
        self.burger_cb = QCheckBox("Burger")
        self.pasta_cb = QCheckBox("Pasta")
        self.salad_cb = QCheckBox("Salad")

        # Setting initial states (optional)
        self.pasta_cb.setChecked(True)

        # Adding checkboxes to layout of this group
        layout.addWidget(self.pizza_cb)
        layout.addWidget(self.burger_cb)
        layout.addWidget(self.pasta_cb)
        layout.addWidget(self.salad_cb)

        # set layout of this group
        self.food_group.setLayout(layout)

    def show_selected_options(self):
        food_options = []
        if self.pizza_cb.isChecked():
            food_options.append("Pizza")
        if self.burger_cb.isChecked():
            food_options.append("Burger")
        if self.pasta_cb.isChecked():
            food_options.append("Pasta")
        if self.salad_cb.isChecked():
            food_options.append("Salad")

        message = "Selected Food Options:\n"
        message += ", ".join(food_options) if food_options else "None"

        QMessageBox.information(self, "Selected Options", message)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = CheckboxOptionsWindow()
    window.show()
    sys.exit(app.exec())
````

##QComboBox

Dropdown Selection

**Add items as a list**

```
self.addItems(["Option 1", "Option 2", "Option 3"])
```

**Set size**

```
self.setMinimumWidth(300)
self.setMinimumHeight(40)
```

**Allow editing**

```
self.setEditable(True)
self.setInsertPolicy(QComboBox.InsertAlphabetically)
```

**Completion mode**

```
self.setCompleter(None)  # Custom completer if needed
```

**Connect signals**

```
self.currentIndexChanged.connect(self.on_selection_changed)

def on_selection_changed(self, index):
  print(f"Selected item: {self.currentText()}")
```

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow,
                             QVBoxLayout, QWidget, QLabel)
from PySide6.QtCore import Qt
from PySide6.QtWidgets import QComboBox

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PySide6 Demo App")
        self.setGeometry(100, 100, 500, 400)

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        # ComboBox - non-editable
        label1 = QLabel("Select Only (Not Editable):")
        non_editable_combo = NonEditableComboBox()

        # ComboBox - editable
        label2 = QLabel("Select or Type (Editable):")
        editable_combo = EditableComboBox()

        # Add to layout
        layout.addWidget(label1)
        layout.addWidget(non_editable_combo)
        layout.addWidget(label2)
        layout.addWidget(editable_combo)
        layout.addStretch()

class NonEditableComboBox(QComboBox):
    def __init__(self):
        super().__init__()

        # Add items
        self.addItems(sorted(["Pizza", "Pasta", "Sushi", "PadThai"]))

        # Non-editable (default is False, but this is for clarity)
        self.setEditable(False)

        # Set size
        self.setMinimumWidth(300)
        self.setMinimumHeight(40)

        # Connect signals
        self.currentIndexChanged.connect(self.on_selection_changed)

    def on_selection_changed(self, index):
        print(f"Non-editable selected: {self.currentText()}")

class EditableComboBox(QComboBox):
    def __init__(self):
        super().__init__()

        # Add items
        self.addItems(sorted(["Pizza", "Pasta", "Sushi", "PadThai"]))

        # Make it editable
        self.setEditable(True)
        self.setInsertPolicy(QComboBox.InsertAlphabetically)
        self.setCompleter(None)

        # Set size
        self.setMinimumWidth(300)
        self.setMinimumHeight(40)

        # Connect signals
        self.currentIndexChanged.connect(self.on_selection_changed)

    def on_selection_changed(self, index):
        print(f"Editable selected/typed: {self.currentText()}")

def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow,
                             QVBoxLayout, QWidget)
from PySide6.QtCore import Qt
from PySide6.QtWidgets import QSpinBox

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("PySide6 Demo App")
        self.setGeometry(100, 100, 400, 500)

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        spinbox = CustomSpinBox()
        layout.addWidget(spinbox, alignment=Qt.AlignCenter)

class CustomSpinBox(QSpinBox):
    def __init__(self):
        super().__init__()

        # Value range and step
        self.setRange(0, 100)
        self.setValue(50)
        self.setSingleStep(5)

        # Prefix and suffix
        self.setPrefix("$")
        self.setSuffix(" units")

        # Display format
        self.setDisplayIntegerBase(10)  # Decimal

        # Connect signals
        self.valueChanged.connect(self.on_value_changed)

    def on_value_changed(self, value):
        print(f"New value: {value}")

def main():
    app = QApplication(sys.argv)
    window = MainWindow()
    window.show()
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

##QMessageBox

- Dialog Windows
- "Pop-up" windows

You don't need a custom class that inherits from QMessageBox
- QMessageBox is a special widget - it's a dialog that Qt lets you use with static methods (QMessageBox.information()) directly
- You don't need to create a custom class for most message boxes
>- using Qt's built-in methods is enough
- There will be examples to using QDialog window (next lecture)


**When you need to inherit:**
- If you want to change the appearance a lot
- If you want to add custom widgets inside the message box

**Information Message**

```
@staticmethod
def show_info(title, message):
    msg = QMessageBox()
    msg.setIcon(QMessageBox.Information)
    msg.setText(message)
    msg.setWindowTitle(title)
    msg.setStandardButtons(QMessageBox.Ok | QMessageBox.Cancel)
    msg.setDefaultButton(QMessageBox.Ok)


    return msg.exec()
```

**Question message**
```
@staticmethod
def show_question(title, question):
    return QMessageBox.question(
        None,
        title,
        question,
        QMessageBox.Yes | QMessageBox.No
    )
```


**Warning message**
```
@staticmethod
def show_warning(title, message):
    return QMessageBox.warning(
        None,
        title,
        message,
        QMessageBox.Ok
```

##QStackedLayout

For switching between widgets/pages

````python
from PySide6.QtWidgets import QStackedLayout, QPushButton

class WizardPages(QWidget):
    def __init__(self):
        super().__init__()
        self.stacked_layout = QStackedLayout()

        # Add pages
        page1 = QWidget()
        page1.setLayout(QVBoxLayout())
        page1.layout().addWidget(QLabel("Page 1 Content"))

        page2 = QWidget()
        page2.setLayout(QVBoxLayout())
        page2.layout().addWidget(QLabel("Page 2 Content"))

        # Add pages to stacked layout
        self.stacked_layout.addWidget(page1)
        self.stacked_layout.addWidget(page2)

        # Navigation buttons
        next_button = QPushButton("Next")
        next_button.clicked.connect(self.next_page)

        # Main layout
        main_layout = QVBoxLayout()
        main_layout.addLayout(self.stacked_layout)
        main_layout.addWidget(next_button)

        self.setLayout(main_layout)

    def next_page(self):
        current = self.stacked_layout.currentIndex()
        self.stacked_layout.setCurrentIndex((current + 1) % self.stacked_layout.count())
````
