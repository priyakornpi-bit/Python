## CSS Style Sheet

- You can style your application using the CSS-like syntax
- Use `(""" """)` block to contain styling

```
widget.setStyleSheet("""
        background-color: #262626;
        color: #FFFFFF;

```

**Ref: https://doc.qt.io/qt-5/stylesheet-reference.html**

````python
import sys
from PySide6.QtCore import Qt
from PySide6.QtWidgets import QApplication, QLabel

if __name__ == "__main__":
    app = QApplication()
    w = QLabel("This is a placeholder text")
    w.setAlignment(Qt.AlignCenter)
    w.setStyleSheet("""
        background-color: #262626;
        color: #FFFFFF;
        font-family: Titillium;
        font-size: 18px;
        """)
    w.show()
    sys.exit(app.exec())
````

## QT Style Sheet

- Styling individual widget one at a time would take time and effort
- Use `qss` file to specify styles and **apply it to the whole application**
- Need to specify the Widget component and optionally the name of the object

For example,
```
QLabel {
    background-color: red;
}

QLabel#title {
    font-size: 20px;
}
```

**Ref: https://doc.qt.io/qt-5/stylesheet-examples.html**

- Include `qss` file in the app folder
- Read the `qss` file
- Then set the style sheet

````python
import sys
import re
from PySide6.QtWidgets import (QApplication, QWidget, QVBoxLayout, QHBoxLayout,
                               QLabel, QListWidget, QListWidgetItem, QPushButton)
from PySide6.QtCore import Qt


class Widget(QWidget):
    def __init__(self, parent=None):
        super(Widget, self).__init__(parent)

        menu_widget = QListWidget()
        for i in range(10):
            item = QListWidgetItem(f"Item {i}")
            item.setTextAlignment(Qt.AlignCenter)
            menu_widget.addItem(item)

        text_widget = QLabel("Hello I am here")
        button = QPushButton("Something")

        content_layout = QVBoxLayout()
        content_layout.addWidget(text_widget)
        content_layout.addWidget(button)
        main_widget = QWidget()
        main_widget.setLayout(content_layout)

        layout = QHBoxLayout()
        layout.addWidget(menu_widget, 1)
        layout.addWidget(main_widget, 4)

        self.setLayout(layout)

if __name__ == "__main__":
    app = QApplication(sys.argv)

    w = Widget()
    w.show()

    with open("style_list.qss", "r") as f:
        _style = f.read()
        app.setStyleSheet(_style)

    sys.exit(app.exec())
````

#Table

## QTableWidget

- Displays data in a spreadsheet-like grid (rows x columns)
- You can customize cell content, colors, and alignment

Ref:
https://doc.qt.io/qtforpython-6/tutorials/basictutorial/tablewidget.html#tutorial-tablewidget


Steps:

1. `from PySide6.QtWidgets import QTableWidget, QTableWidgetItem`

2. Create a table object and define table dimensions

`mytable = QTableWidget(rows, cols)`
or
```
mytable = QTableWidget()
mytable.self.table.setColumnCount(cols)
```

3. Set column headers

`mytable.setHorizontalHeaderLabels()`

4. populate each cell

`setItem(row, col, QTableWidgetItem(text))`:

- Use loop!
- Every value must be QTableWidgetItem(str)

5. Set other properties, e.g.
>- change cell background color using `setBackground(QColor(...))`

###QTableWidgetItem

- How Qt stores cell data along with extra properties like color, font, alignment, etc.
- So you can do these later using QTableWidgetItem()
```
item = QTableWidgetItem("Alice")
item.setBackground(QColor("yellow"))    # background color
item.setForeground(QColor("red"))       # text color
item.setTextAlignment(Qt.AlignCenter)   # alignment
```

````python
import sys
from PySide6.QtGui import QColor
from PySide6.QtWidgets import (QApplication, QTableWidget,
                               QTableWidgetItem, QMainWindow)

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("QTableWidget Example")

        # Create a table with 4 rows and 3 columns
        table = QTableWidget(4, 3)
        table.setHorizontalHeaderLabels(["Name", "Age", "Score"])

        # Populate with data
        data = [
            ("Alice", 25, 95),
            ("Bob", 30, 82),
            ("Charlie", 22, 78),
            ("Diana", 28, 91),
        ]

        for row, (name, age, score) in enumerate(data):
            table.setItem(row, 0, QTableWidgetItem(name))
            table.setItem(row, 1, QTableWidgetItem(str(age)))

            # Color-code scores
            score_item = QTableWidgetItem(str(score))
            if score >= 90:
                score_item.setBackground(QColor("#2ABf9E"))  # green
            else:
                score_item.setBackground(QColor("#FF6B6B"))  # red
            table.setItem(row, 2, score_item)

        self.setCentralWidget(table)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MainWindow()
    window.resize(400, 300)
    window.show()
    sys.exit(app.exec())
````

**Data enumeration**

`enumerate()` adds a counter to each item when looping through a list.

Data
```
data = [
            ("Alice", 25, 95),
            ("Bob", 30, 82),
            ("Charlie", 22, 78),
            ("Diana", 28, 91),
        ]
```

Enumeration
```
for row, (name, age, score) in enumerate(data):
    table.setItem(row, 0, QTableWidgetItem(name))
    table.setItem(row, 1, QTableWidgetItem(str(age)))
    table.setItem(row, 2, QTableWidgetItem(str(score)))
```

vs

Simple loop
```
for i in range(len(data)):
    name = data[i][0]
    age = data[i][1]
    score = data[i][2]
```

## PyQtGraph

- High-performance plotting library built on top of Qt
- Ideal for real-time data visualization because it renders directly as a Qt widget with no bridging needed

Ref:
https://www.pythonguis.com/tutorials/pyside6-plotting-pyqtgraph/

Steps:

First, install the module (one time)

```
python.exe -m pip install --upgrade pip
pip install pyqtgraph
```


1. `import pyqtgraph as pg`

2. Create a plot

`plot_widget = pg.PlotWidget()`

3. Set plot properties, e.g.

- Set background

`plot_widget.setBackground("w")`

- Set title

`plot_widget.setTitle("Sin & Cos Waves", color="b", size="16pt")`

- Set y-axis label

`plot_widget.setLabel("left", "Amplitude")`

- Set x-axis label

`plot_widget.setLabel("bottom", "Time (s)")`

- Set legend

`plot_widget.addLegend()` or <br>
`plot_widget.addLegend(offset=(70, 30))` to change the legend position

4. Get data

- use numpy to generate math function data
- use your own data

5. Actually put line on the plot

`plot_widget.plot(x, y, line_properties, legend_name)`

`plot_widget.plot(x, y_sin, pen=pg.mkPen("r", width=2), name="sin(x)")`

6. You can also plot multiple lines by calling plot() more than once

```
plot_widget.plot(months, sales_a, pen=pg.mkPen("b", width=2), name="Store A")
plot_widget.plot(months, sales_b, pen=pg.mkPen("r", width=2), name="Store B")
```


**Plotting from math functions**

````python
import sys
import numpy as np
from PySide6.QtWidgets import QApplication, QMainWindow
import pyqtgraph as pg

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("PyQtGraph Example")

        # Create a PlotWidget
        plot_widget = pg.PlotWidget()
        plot_widget.setBackground("w")
        plot_widget.setTitle("Sin & Cos Waves", color="b", size="16pt")
        plot_widget.setLabel("left", "Amplitude")
        plot_widget.setLabel("bottom", "Time (s)")
        plot_widget.addLegend(offset=(30, 200))

        # Generate data
        x = np.linspace(0, 10, 100)
        y_sin = np.sin(x)
        y_cos = np.cos(x)

        # See what x, y_sin, y_cos give us
        print(x)
        print(y_sin)
        print(y_cos)

        # Plot lines
        plot_widget.plot(x, y_sin, pen=pg.mkPen("r", width=2), name="sin(x)")
        plot_widget.plot(x, y_cos, pen=pg.mkPen("b", width=2,
                         style=pg.QtCore.Qt.DashLine), name="cos(x)")

        self.setCentralWidget(plot_widget)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MainWindow()
    window.resize(600, 400)
    window.show()
    sys.exit(app.exec())
````

**Plotting from data**

````python
import sys
from PySide6.QtWidgets import QApplication, QMainWindow
import pyqtgraph as pg

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Sales Chart")

        # Your data
        months = [1, 2, 3, 4, 5, 6]
        sales_a = [150, 200, 180, 250, 300, 280]
        sales_b = [100, 120, 160, 200, 220, 260]

        # Create plot
        plot_widget = pg.PlotWidget()
        plot_widget.setBackground("w")
        plot_widget.setTitle("Monthly Sales")
        plot_widget.setLabel("left", "Sales ($)")
        plot_widget.setLabel("bottom", "Month")
        plot_widget.addLegend()

        # Plot the data
        plot_widget.plot(months, sales_a, pen=pg.mkPen("b", width=2), name="Store A")
        plot_widget.plot(months, sales_b, pen=pg.mkPen("r", width=2), name="Store B")

        self.setCentralWidget(plot_widget)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MainWindow()
    window.resize(600, 400)
    window.show()
    sys.exit(app.exec())
````
