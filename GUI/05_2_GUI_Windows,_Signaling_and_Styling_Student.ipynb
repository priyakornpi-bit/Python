## QMainWindow - โครงสร้างพื้นฐาน

### การตั้งค่า Central Widget

**Central widget** คือพื้นที่เนื้อหาหลักที่คุณวาง UI หลักของคุณ

```python
central_widget = QWidget()
self.setCentralWidget(central_widget)
```

⚠️ **สำคัญ:**
- **ถ้าไม่มี central widget คุณจะไม่สามารถเพิ่มวิดเจ็ตไปที่พื้นที่หลักของหน้าต่างได้**
- คุณไม่สามารถใช้ layout โดยตรงกับ QMainWindow ได้
  - ต้องใช้ layout กับ central widget แทน

**ตัวอย่าง:**
```python
# สร้าง central widget
central_widget = QWidget()
self.setCentralWidget(central_widget)

# สร้าง layout สำหรับ central widget
layout = QVBoxLayout(central_widget)

# เพิ่มวิดเจ็ตเข้าไปใน layout
layout.addWidget(button)
```


````python
from PySide6.QtWidgets import QApplication, QMainWindow, QWidget, QPushButton, QVBoxLayout
import sys

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        # Basic window setup
        self.setWindowTitle("My App")
        self.setGeometry(100, 100, 800, 600)  # x, y, width, height

        # -------------------------------------------------
        # IMPORTANT: QMainWindow needs a central widget
        # -------------------------------------------------
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        # -------------------------------------------------

        # Create layout for central widget
        layout = QVBoxLayout(central_widget)

        # Add widgets to layout
        button = QPushButton("Click Me")
        layout.addWidget(button)
        button = QPushButton("Click Me Too!")
        layout.addWidget(button)
        button = QPushButton("Click Me Three \^^/")
        layout.addWidget(button)

def main():
    # Create the Qt Application (reuse existing instance if available)
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    # Create and show the main window
    window = MainWindow()
    window.show()

    # Start the event loop
    sys.exit(app.exec())

if __name__ == "__main__":
    main()
````

## หน้าต่างแบบธรรมดาด้วย QWidget


### QWidget ในฐานะหน้าต่าง

**เมื่อใช้ QWidget เป็นหน้าต่าง:**
- เมื่อ QWidget ไม่มี parent (หรือกำหนดเป็น window อย่างชัดเจน) มันจะกลายเป็นหน้าต่างอิสระ
- มันจะได้กรอบหน้าต่าง, แถบชื่อเรื่อง และเมนูระบบของตัวเอง
- สามารถเคลื่อนย้าย, ปรับขนาด, ย่อเล็กสุด และปิดได้อย่างอิสระ
- สามารถใช้ layout โดยตรงกับ QWidget ได้ (ต่างจาก QMainWindow)

**กรณีการใช้งานทั่วไป:**
- หน้าต่างเครื่องมือหรืออรรถประโยชน์แบบง่าย
- หน้าจอ splash แบบกำหนดเอง
- แผงหรือจานสีแบบลอย
- หน้าต่างรองที่มีน้ำหนักเบา

**QWidget ในฐานะคอนเทนเนอร์:**
เมื่อ QWidget มี parent มันจะทำหน้าที่เป็นวิดเจ็ตภายใน parent นั้น:
- กลายเป็นส่วนหนึ่งของ layout ของ parent
- ถูกจำกัดอยู่ภายในขอบเขตของ parent
- ไม่สามารถเคลื่อนออกนอกขอบเขตของ parent
- ถูกตัด (cut off) ถ้าพยายามวาดนอกพื้นที่ของ parent
- สืบทอดกฎการมองเห็นจาก parent (ถ้า parent ถูกซ่อน child ก็ถูกซ่อนด้วย)


````python
from PySide6.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel
import sys

# Inherit from QWidget = using QWidget as a "Window"
class SimpleWindow(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Simple Window")

        # QWidget can directly use layouts
        layout = QVBoxLayout(self)

        label = QLabel("Hello World") # widget as a widget
        layout.addWidget(label)

if __name__ == '__main__':
    # Create the application instance (reuse existing if available)
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    # Create and show the window
    window = SimpleWindow() # QWidget as a window
    window.show()

    # Start the event loop
    sys.exit(app.exec())
````

# ระบบไดอะล็อก

## QDialog - หน้าต่างแบบโมดอลและโมดเลส

**วัตถุประสงค์:**
- QDialog มีไว้สำหรับ**หน้าต่างชั่วคราว**
- ใช้สำหรับการโต้ตอบกับผู้ใช้ระยะสั้น

**กรณีการใช้งานทั่วไป:**
- ไดอะล็อกการตั้งค่าคอนฟิก
- ไดอะล็อกการดำเนินการไฟล์
- ไดอะล็อกยืนยัน
- ข้อความแสดงข้อผิดพลาด
- การรับข้อมูลจากผู้ใช้
- กล่องเกี่ยวกับ

**ประเภทของไดอะล็อก:**

### ไดอะล็อกแบบโมดอล
```python
result = dialog.exec()  # บล็อกการโต้ตอบกับหน้าต่าง parent
```
- บล็อกการโต้ตอบกับหน้าต่าง parent จนกว่าจะปิด
- ผู้ใช้ต้องจัดการกับไดอะล็อกก่อนกลับไปที่หน้าต่างหลัก
- คืนค่ารหัสผลลัพธ์ (Accepted หรือ Rejected)

### ไดอะล็อกแบบโมดเลส
```python
dialog.show()  # อนุญาตให้โต้ตอบกับหน้าต่าง parent ได้
```
- อนุญาตให้โต้ตอบกับหน้าต่าง parent ขณะที่ไดอะล็อกเปิดอยู่
- ยืดหยุ่นกว่าแต่ต้องจัดการแตกต่างกัน

**ผลลัพธ์ของไดอะล็อก:**
- `accept()` - ปิดไดอะล็อกด้วย `QDialog.Accepted (1)`
- `reject()` - ปิดไดอะล็อกด้วย `QDialog.Rejected (0)`

**ความสัมพันธ์แบบ Parent-Child:**
- ไดอะล็อกมักจะมีหน้าต่าง parent
- เมื่อ parent ถูกปิด ไดอะล็อกจะถูกปิดโดยอัตโนมัติ
- ไดอะล็อกจะปรากฏตรงกลางเหนือหน้าต่าง parent


### QDialogButtonBox - ปุ่มไดอะล็อกมาตรฐาน

ชุดปุ่มที่กำหนดไว้สำหรับไดอะล็อก แทนที่จะสร้างปุ่มด้วยตนเอง ให้ใช้ชุดปุ่มมาตรฐาน:

**ชุดปุ่มทั่วไป:**
- Ok, Cancel
- Yes, No
- Apply, Reset, Help
- SaveAll, Save, Close, Open
- Retry, Ignore, Discard, RestoreDefaults

**ประโยชน์:**
- UI ที่สอดคล้องกันในทุกแพลตฟอร์ม
- จัดการการจัดเรียงปุ่มโดยอัตโนมัติ
- พฤติกรรมมาตรฐานที่ผู้ใช้คาดหวัง

**เอกสารอ้างอิง:**
- https://doc.qt.io/qtforpython-6/PySide6/QtWidgets/QDialogButtonBox.html
- https://doc.qt.io/qtforpython-6/PySide6/QtWidgets/QDialogButtonBox.html#PySide6.QtWidgets.QDialogButtonBox.StandardButton


````python
from PySide6.QtWidgets import QApplication, QDialog, QVBoxLayout, QLabel, QDialogButtonBox
import sys

class MyDialog(QDialog):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Dialog Example")

        # Create layout
        layout = QVBoxLayout(self)

        # Add some content
        label = QLabel("Do you want to save changes?")
        layout.addWidget(label)

        # Create QDialogButtonBox with standard buttons
        button_box = QDialogButtonBox(
            QDialogButtonBox.Ok | QDialogButtonBox.Cancel
        )

        # Connect signals to slots
        button_box.accepted.connect(self.on_accept)  # OK button clicked
        button_box.rejected.connect(self.on_reject)  # Cancel button clicked

        layout.addWidget(button_box)

    def on_accept(self):
        """Called when OK button is clicked"""
        print("OK button clicked - Saving changes...")
        self.accept()  # Close dialog with Accepted status

    def on_reject(self):
        """Called when Cancel button is clicked"""
        print("Cancel button clicked - Discarding changes...")
        self.reject()  # Close dialog with Rejected status

if __name__ == '__main__':
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    dialog = MyDialog()
    result = dialog.exec()  # Show dialog and wait for response

    # Check the result after dialog closes
    if result == QDialog.Accepted:
        print("Dialog result: Accepted")
    else:
        print("Dialog result: Rejected")

    sys.exit()
````

### QMessageBox - ข้อความ, คำเตือน และคำถาม

สำหรับแสดงข้อความ, คำเตือน และคำถามให้กับผู้ใช้

**ข้อความแจ้งข่าว**

```python
QMessageBox.information(None, "หัวข้อ", "นี่คือข้อความแจ้งข่าว")
```

**ข้อความเตือน**
```python

```

**ข้อผิดพลาดร้ายแรง**
```python

```

**คำถามพร้อมปุ่ม Yes/No**
```python
reply = QMessageBox.question(None, "คำถาม",
        "คุณต้องการดำเนินการต่อหรือไม่?",
        QMessageBox.Yes | QMessageBox.No)

if reply == QMessageBox.Yes:
    # จัดการคำตอบ Yes
else:
    # จัดการคำตอบ No
```


````python
import sys
from PySide6.QtWidgets import QApplication, QMessageBox, QWidget

class MessageBoxDemo(QWidget):
    def __init__(self):
        super().__init__()

        # Show message boxes immediately when application starts
        self.show_message_boxes()

    def show_message_boxes(self):
        # Information message
        QMessageBox.information(self, "Title", "This is an information message")

        # Warning message
        QMessageBox.warning(self, "Warning", "This is a warning message")

        # Critical error
        QMessageBox.critical(self, "Error", "This is an error message")

        # Question with Yes/No buttons
        reply = QMessageBox.question(self, "Question",
                                  "Do you want to proceed?",
                                  QMessageBox.Yes | QMessageBox.No)
        if reply == QMessageBox.Yes:
            print("User clicked Yes")
        else:
            print("User clicked No")

def main():
    # Create the application (reuse existing if available)
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    # Create and show the main window
    demo = MessageBoxDemo()

    # Start the event loop
    #sys.exit(app.exec())
    # Force quit instead
    app.quitOnLastWindowClosed()

if __name__ == "__main__":
    main()
````

````python
import sys
from PySide6.QtWidgets import QApplication, QWidget
from PySide6.QtWidgets import QFileDialog

class FileDemo(QWidget):
    def __init__(self):
        super().__init__()

        # Show message boxes immediately when application starts
        self.show_file_selection()

    def show_file_selection(self):
        # Open file dialog
        filename, _ = QFileDialog.getOpenFileName(
            None,
            "Open File",
            "",  # Starting directory
            "Text Files (*.txt);;All Files (*)"  # File filters
        )

        # Save file dialog
        filename, _ = QFileDialog.getSaveFileName(
            None,
            "Save File",
            "",
            "Text Files (*.txt)"
        )

        # Select directory
        directory = QFileDialog.getExistingDirectory(
            None,
            "Select Directory"
        )


def main():
    # Create the application (reuse existing if available)
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    # Create and show the main window
    demo = FileDemo()

    # Start the event loop
    #sys.exit(app.exec())
    # Force quit instead
    app.quitOnLastWindowClosed()

if __name__ == "__main__":
    main()
````

## ไดอะล็อกแบบกำหนดเอง

สร้างกล่องไดอะล็อกของคุณเองสำหรับกรณีการใช้งานพิเศษ

**ขั้นตอนการสร้างไดอะล็อกแบบกำหนดเอง:**
1. สร้างคลาส**สืบทอดจาก QDialog**
2. สร้าง layout สำหรับไดอะล็อก
3. เพิ่มวิดเจ็ตแบบกำหนดเองของคุณเข้าไปใน layout
4. เชื่อมต่อปุ่ม OK/Cancel กับสล็อต `accept()` และ `reject()`
5. เข้าถึงค่าวิดเจ็ตหลังจาก `exec()` คืนค่า

**โครงสร้างตัวอย่าง:**
```python
class CustomDialog(QDialog):
    def __init__(self, parent=None):
        super().__init__(parent)
        self.setWindowTitle("ไดอะล็อกของฉัน")
        
        # สร้าง layout
        layout = QFormLayout(self)
        
        # เพิ่มวิดเจ็ต
        self.name_input = QLineEdit()
        layout.addRow("ชื่อ:", self.name_input)
        
        # เพิ่มปุ่ม
        buttons = QDialogButtonBox(
            QDialogButtonBox.Ok | QDialogButtonBox.Cancel
        )
        buttons.accepted.connect(self.accept)
        buttons.rejected.connect(self.reject)
        layout.addRow(buttons)
```

**การใช้ไดอะล็อกแบบกำหนดเอง:**
```python
dialog = CustomDialog()
if dialog.exec() == QDialog.Accepted:
    print("ยอมรับ!")
    print(dialog.name_input.text())
else:
    print("ปฏิเสธ!")
```


````python
from PySide6.QtWidgets import (QApplication, QDialog, QFormLayout,
                             QLineEdit, QDialogButtonBox)
import sys

# Custom QDialog window
class CustomDialog(QDialog):
   def __init__(self, parent=None):
        super().__init__(parent)
        self.setWindowTitle("Dialog")

        # -------------------------------------------------------
        # Notice: this is not using the QDialog properties
        # unlike the prev examples
        # -------------------------------------------------------
        # Dialogs often use form layouts
        layout = QFormLayout(self)

        self.name_input = QLineEdit()
        layout.addRow("Name:", self.name_input)

        # Dialog buttons
        buttons = QDialogButtonBox(
            QDialogButtonBox.Ok | QDialogButtonBox.Cancel
        )
        buttons.accepted.connect(self.accept)
        buttons.rejected.connect(self.reject)
        layout.addRow(buttons)
        # -------------------------------------------------------

if __name__ == '__main__':
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    
    dialog = CustomDialog()

    # Show dialog and get result
    result = dialog.exec()
    if result:
        print("Dialog accepted!")
        print("Name entered:", dialog.name_input.text())
    else:
        print("Dialog rejected!")

    # no need for app.exec here,
    # because there is only the dialog window opening for this application
    # Force quit here
    app.quitOnLastWindowClosed()
````

## ตัวอย่างไดอะล็อก {#dialog-example}

ตัวอย่างที่ใช้งานได้จริงแบบสมบูรณ์พร้อมไดอะล็อกทุกประเภทรวมอยู่ในแอปพลิเคชันเดียว

**คุณสมบัติของแอปพลิเคชัน:**
- เป็นโมดอลโดยค่าเริ่มต้น (บล็อกการโต้ตอบกับหน้าต่าง parent)
- คืนค่าสถานะ (มักเป็นบูลีน) ที่บ่งชี้ว่าผู้ใช้ยอมรับหรือยกเลิก
- จัดการ layout และการตรวจสอบของตัวเอง
- ทำตามลักษณะเนทีฟของระบบ

**โครงสร้างตัวอย่าง:**
โปรแกรมมี MainWindow ที่ประกอบด้วยปุ่มสำหรับแต่ละประเภทของไดอะล็อก เมื่อคุณคลิกปุ่ม แอปพลิเคชันจะเปิดไดอะล็อกที่สอดคล้องกัน ดูตัวอย่างแบบสมบูรณ์ด้านล่าง


````python
from PySide6.QtWidgets import (QApplication, QMainWindow, QVBoxLayout,
                              QWidget, QPushButton, QMessageBox, QFileDialog,
                              QInputDialog, QColorDialog, QFontDialog)
from PySide6.QtGui import QColor, QFont
from PySide6.QtCore import Qt
import sys

class DialogDemo(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Dialog Types Demo")
        self.setGeometry(100, 100, 400, 500)

        # Central widget and layout
        central = QWidget()
        self.setCentralWidget(central)
        layout = QVBoxLayout(central)

        # Create buttons for each dialog type
        # Message Boxes
        layout.addWidget(QPushButton("Information", clicked=self.show_information))
        layout.addWidget(QPushButton("Warning", clicked=self.show_warning))
        layout.addWidget(QPushButton("Critical", clicked=self.show_critical))
        layout.addWidget(QPushButton("Question", clicked=self.show_question))

        # File Dialogs
        layout.addWidget(QPushButton("Open File", clicked=self.show_open_file))
        layout.addWidget(QPushButton("Save File", clicked=self.show_save_file))
        layout.addWidget(QPushButton("Select Directory", clicked=self.show_directory))

        # Input Dialogs
        layout.addWidget(QPushButton("Text Input", clicked=self.show_text_input))
        layout.addWidget(QPushButton("Integer Input", clicked=self.show_int_input))
        layout.addWidget(QPushButton("Double Input", clicked=self.show_double_input))
        layout.addWidget(QPushButton("Item Input", clicked=self.show_item_input))

        # Color Dialog
        layout.addWidget(QPushButton("Color Picker", clicked=self.show_color_dialog))

        # Font Dialog
        layout.addWidget(QPushButton("Font Picker", clicked=self.show_font_dialog))

        # Done button
        done_button = QPushButton("Done")
        done_button.setFixedSize(100, 40)
        done_button.move(150, 80)
        done_button.clicked.connect(self.close)
        layout.addWidget(done_button, alignment=Qt.AlignRight)

    # Done
    def done(self):
        pass

    # Message Box methods
    def show_information(self):
        QMessageBox.information(
            self,
            "Information",
            "This is an information message."
        )

    def show_warning(self):
        QMessageBox.warning(
            self,
            "Warning",
            "This is a warning message!"
        )

    def show_critical(self):
        QMessageBox.critical(
            self,
            "Error",
            "This is a critical error message!"
        )

    def show_question(self):
        reply = QMessageBox.question(
            self,
            "Question",
            "Would you like to proceed?",
            QMessageBox.Yes | QMessageBox.No | QMessageBox.Cancel
        )

        if reply == QMessageBox.Yes:
            print("User clicked Yes")
        elif reply == QMessageBox.No:
            print("User clicked No")
        else:
            print("User clicked Cancel")

    # File Dialog methods
    def show_open_file(self):
        filename, selected_filter = QFileDialog.getOpenFileName(
            self,
            "Open File",
            "",  # Start in current directory
            "Text Files (*.txt);;Python Files (*.py);;All Files (*)"
        )
        if filename:
            print(f"Selected file to open: {filename}")

    def show_save_file(self):
        filename, selected_filter = QFileDialog.getSaveFileName(
            self,
            "Save File",
            "",
            "Text Files (*.txt);;Python Files (*.py)"
        )
        if filename:
            print(f"Selected file to save: {filename}")

    def show_directory(self):
        directory = QFileDialog.getExistingDirectory(
            self,
            "Select Directory",
            "",
            QFileDialog.ShowDirsOnly
        )
        if directory:
            print(f"Selected directory: {directory}")

    # Input Dialog methods
    def show_text_input(self):
        text, ok = QInputDialog.getText(
            self,
            "Text Input",
            "Enter your name:",
            text="Default Text"
        )
        if ok:
            print(f"Text entered: {text}")

    def show_int_input(self):
        number, ok = QInputDialog.getInt(
            self,
            "Integer Input",
            "Enter your age:",
            value=25,  # Default value
            min=0,
            max=150,
            step=1
        )
        if ok:
            print(f"Number entered: {number}")

    def show_double_input(self):
        number, ok = QInputDialog.getDouble(
            self,
            "Double Input",
            "Enter amount:",
            value=0.0,
            min=-10000.0,
            max=10000.0,
            decimals=2
        )
        if ok:
            print(f"Amount entered: {number}")

    def show_item_input(self):
        items = ["Item 1", "Item 2", "Item 3", "Item 4"]
        item, ok = QInputDialog.getItem(
            self,
            "Item Selection",
            "Select an item:",
            items,
            current=0,  # Default selected item index
            editable=False
        )
        if ok:
            print(f"Selected item: {item}")

    # Color Dialog method
    def show_color_dialog(self):
        initial = QColor('red')
        color = QColorDialog.getColor(
            initial,
            self,
            "Select Color"
        )
        if color.isValid():
            print(f"Selected color: {color.name()}")
            # You could use this color to set background of a widget:
            # self.setStyleSheet(f"background-color: {color.name()}")

    # Font Dialog method
    def show_font_dialog(self):
        initial = QFont("Arial", 12)
        ok, font = QFontDialog.getFont(
            initial,
            self,
            "Select Font"
        )
        if ok:
            print(f"Selected font: {font.family()}")
            print(f"Font size: {font.pointSize()}")
            # You could use this font for a widget:
            # self.setFont(font)

if __name__ == '__main__':
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    window = DialogDemo()
    window.show()
    sys.exit(app.exec())
````

# สัญญาณและสล็อต

กลไกหลักของการเขียนโปรแกรมแบบ event-driven ใน Qt

**แนวคิด:**
- **สัญญาณ (Signal)** - การโต้ตอบที่ผู้ใช้ทำกับวิดเจ็ต (เช่น คลิก, พิมพ์, เปลี่ยนค่า)
- **สล็อต (Slot)** - แอคชัน/ฟังก์ชันที่เกิดขึ้นเมื่อสัญญาณถูกส่งออก (เช่น บันทึกไฟล์, อัพเดตป้ายชื่อ)

**ตัวอย่าง:**
ผู้ใช้คลิกปุ่มเพื่อบันทึกไฟล์:
- **สัญญาณ**: `button.clicked`
- **สล็อต**: ฟังก์ชัน `save_file()`

**ทำไมต้องใช้สัญญาณและสล็อต?**
- แยก UI ออกจาก business logic
- ทำให้โค้ดดูแลรักษาง่ายขึ้น
- อนุญาตให้มีหลายแอคชันจากอีเวนต์เดียว
- ง่ายต่อการเชื่อมต่อ/ตัดการเชื่อมต่อตอนรันไทม์

**เอกสารอ้างอิง:** https://doc.qt.io/qtforpython-6/tutorials/basictutorial/signals_and_slots.html


## การเชื่อมต่อสัญญาณและสล็อต

### การเชื่อมต่อพื้นฐาน

ใช้**เมธอด `connect()`** เพื่อเชื่อมต่อสัญญาณกับสล็อต:

```python
widget.signal.connect(slot_function)
```

**ประเภทการเชื่อมต่อ:**
- `Qt.AutoConnection` - ค่าเริ่มต้น, ประเภทการเชื่อมต่อที่เหมาะสมที่สุด
- `Qt.DirectConnection` - เรียกทันที
- `Qt.QueuedConnection` - เรียกแบบเลื่อน
- `Qt.BlockingQueuedConnection` - รอการดำเนินการ

**เอกสารอ้างอิง:** https://doc.qt.io/qtforpython-6/PySide6/QtCore/Qt.html#PySide6.QtCore.Qt.ConnectionType

### ตัวอย่าง

**เชื่อมต่อการคลิกปุ่มกับสล็อตในตัว**
```python
button = QPushButton("ล้าง")
line_edit = QLineEdit()

# เชื่อมต่อสัญญาณกับสล็อตในตัว
button.clicked.connect(line_edit.clear)
```

**เชื่อมต่อการคลิกปุ่มกับฟังก์ชันแบบกำหนดเอง**
```python
def on_button_clicked():
    print("ปุ่มถูกคลิก!")

button = QPushButton("คลิกฉัน")
button.clicked.connect(on_button_clicked)
```

**เชื่อมต่อด้วย Lambda (ส่งพารามิเตอร์)**
```python
def on_button_clicked(name):
    print(f"สวัสดี, {name}!")

button = QPushButton("คลิกฉัน")
button.clicked.connect(lambda: on_button_clicked("ผู้ใช้"))
```

### จุดสำคัญ

- คุณสามารถเชื่อมต่อ**สัญญาณได้มากเท่าที่ต้องการไปยังสล็อตเดียว**
- คุณสามารถเชื่อมต่อ**สล็อตได้มากเท่าที่ต้องการไปยังสัญญาณเดียว**
- ใช้ฟังก์ชัน lambda เพื่อส่งพารามิเตอร์ไปยังสล็อต
- สล็อตในตัว: `setText()`, `clear()`, `setEnabled()`, ฯลฯ


- You can connect **as many signals** as you want **to a single slot**
- You can connect **as many slots** as you need **to a signal**

### หมายเหตุ: ฟังก์ชัน Lambda

Lambda ใน Python เป็นวิธีการสร้างฟังก์ชันแบบไม่มีชื่อขนาดเล็กในบรรทัดเดียว

**ไวยากรณ์:**
```python
lambda arguments: expression
```

**ตัวอย่าง:**

ใช้ lambda:
```python
add = lambda x, y: x + y
result = add(3, 5)  # คืนค่า 8
```

ฟังก์ชันปกติที่เทียบเท่า:
```python
def add(x, y):
    return x + y
result = add(3, 5)  # คืนค่า 8
```

**การจัดการพารามิเตอร์ของสัญญาณ:**

````python
from PySide6.QtWidgets import QApplication, QWidget, QVBoxLayout, QLabel, QPushButton, QLineEdit, QCheckBox, QRadioButton, QComboBox, QDateEdit, QSpinBox, QMessageBox, QTextEdit
from PySide6.QtCore import QDate
import sys

class WidgetDemo(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("PySide6 Multiple Signals Demo")
        layout = QVBoxLayout()

        # QLabel - 2 signals
        self.label = QLabel('<a href="http://example.com">Hover or click this link</a>')
        self.label.setOpenExternalLinks(False)
        self.label.linkActivated.connect(lambda url: print(f"Link clicked: {url}"))
        self.label.linkHovered.connect(lambda url: print(f"Link hovered: {url}"))
        layout.addWidget(self.label)

        # QPushButton - 4 signals
        btn = QPushButton("Press Me")
        btn.clicked.connect(lambda: print("Button clicked!"))
        btn.pressed.connect(lambda: print("Button pressed down"))
        btn.released.connect(lambda: print("Button released"))
        btn.setCheckable(True)
        btn.toggled.connect(lambda checked: print(f"Button toggled: {checked}"))
        layout.addWidget(btn)

        # QLineEdit - 5 signals
        line_edit = QLineEdit()
        line_edit.setPlaceholderText("Type and press Enter...")
        line_edit.textChanged.connect(lambda text: print(f"Text changed: {text}"))
        line_edit.textEdited.connect(lambda text: print(f"Text edited by user: {text}"))
        line_edit.returnPressed.connect(lambda: print("Enter key pressed!"))
        line_edit.editingFinished.connect(lambda: print("Editing finished"))
        line_edit.selectionChanged.connect(lambda: print("Selection changed"))
        layout.addWidget(line_edit)

        # QCheckBox - 3 signals
        checkbox = QCheckBox("Check me")
        checkbox.stateChanged.connect(lambda state: print(f"Checkbox state: {state}"))
        checkbox.toggled.connect(lambda checked: print(f"Checkbox toggled: {checked}"))
        checkbox.clicked.connect(lambda checked: print(f"Checkbox clicked: {checked}"))
        layout.addWidget(checkbox)

        # QRadioButton - 2 signals
        radio = QRadioButton("Select me")
        radio.toggled.connect(lambda checked: print(f"Radio toggled: {checked}"))
        radio.clicked.connect(lambda checked: print(f"Radio clicked: {checked}"))
        layout.addWidget(radio)

        # QComboBox - 4 signals
        combo = QComboBox()
        combo.addItems(["Option 1", "Option 2", "Option 3"])
        combo.currentIndexChanged.connect(lambda idx: print(f"Index changed: {idx}"))
        combo.currentTextChanged.connect(lambda text: print(f"Text changed: {text}"))
        combo.activated.connect(lambda idx: print(f"User activated index: {idx}"))
        combo.textActivated.connect(lambda text: print(f"User activated: {text}"))
        layout.addWidget(combo)

        # QDateEdit - 2 signals
        date_edit = QDateEdit()
        date_edit.setDate(QDate.currentDate())
        date_edit.dateChanged.connect(lambda date: print(f"Date changed: {date.toString()}"))
        date_edit.dateTimeChanged.connect(lambda dt: print(f"DateTime changed: {dt.toString()}"))
        layout.addWidget(date_edit)

        # QSpinBox - 2 signals
        spinbox = QSpinBox()
        spinbox.valueChanged.connect(lambda val: print(f"SpinBox value: {val}"))
        spinbox.textChanged.connect(lambda text: print(f"SpinBox text: {text}"))
        layout.addWidget(spinbox)

        # QMessageBox (triggered by button)
        msg_btn = QPushButton("Show MessageBox")
        msg_btn.clicked.connect(self.show_message)
        layout.addWidget(msg_btn)

        # QTextEdit - 5 signals
        text_edit = QTextEdit()
        text_edit.setPlaceholderText("Type here...")
        text_edit.textChanged.connect(lambda: print("TextEdit: text changed"))
        text_edit.selectionChanged.connect(lambda: print("TextEdit: selection changed"))
        text_edit.cursorPositionChanged.connect(lambda: print("TextEdit: cursor moved"))
        text_edit.undoAvailable.connect(lambda available: print(f"Undo available: {available}"))
        text_edit.redoAvailable.connect(lambda available: print(f"Redo available: {available}"))
        layout.addWidget(text_edit)

        self.setLayout(layout)

    def show_message(self):
        msg = QMessageBox()
        msg.setText("Hello from QMessageBox!")
        msg.setStandardButtons(QMessageBox.Ok | QMessageBox.Cancel)
        msg.buttonClicked.connect(lambda btn: print(f"MessageBox button: {btn.text()}"))
        msg.exec()

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    window = WidgetDemo()
    window.resize(400, 600)
    window.show()
    sys.exit(app.exec())
````

ตัวอย่างอื่น


````python
import sys
from PySide6.QtWidgets import (QApplication, QMainWindow, QWidget,
                             QVBoxLayout, QPushButton, QLabel, QSpinBox,
                             QLineEdit, QProgressBar, QCheckBox)

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Signals and Slots Demo")

        # Create central widget and layout
        central_widget = QWidget()
        self.setCentralWidget(central_widget)
        layout = QVBoxLayout(central_widget)

        # Create widgets
        self.counter_label = QLabel("Counter: 0")
        self.increment_button = QPushButton("Increment")
        self.reset_button = QPushButton("Reset")

        # Progress bar example
        self.progress = QProgressBar()
        self.progress.setRange(0, 100)
        self.spin_box = QSpinBox()
        self.spin_box.setRange(0, 100)

        # Text input example
        self.text_input = QLineEdit()
        self.text_label = QLabel("Enter text above")

        # Checkbox and button example
        self.checkbox = QCheckBox("Enable Button")
        self.action_button = QPushButton("Click Me!")
        self.action_button.setEnabled(False)

        # Add widgets to layout
        layout.addWidget(self.counter_label)
        layout.addWidget(self.increment_button)
        layout.addWidget(self.reset_button)
        layout.addWidget(self.progress)
        layout.addWidget(self.spin_box)
        layout.addWidget(self.text_input)
        layout.addWidget(self.text_label)
        layout.addWidget(self.checkbox)
        layout.addWidget(self.action_button)

        # Initialize counter
        self.counter = 0

        # Connect signals to slots
        self.increment_button.clicked.connect(self.increment_counter)
        self.reset_button.clicked.connect(self.reset_counter)
        self.spin_box.valueChanged.connect(self.progress.setValue)
        self.text_input.textChanged.connect(self.text_label.setText)
        self.checkbox.toggled.connect(self.action_button.setEnabled)
        self.action_button.clicked.connect(self.increment_counter)

    def increment_counter(self):
        self.counter += 1
        self.counter_label.setText(f"Counter: {self.counter}")

    def reset_counter(self):
        self.counter = 0
        self.counter_label.setText("Counter: 0")
        self.progress.setValue(0)
        self.spin_box.setValue(0)
        self.text_input.clear()
        self.checkbox.setChecked(False)

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    window = MainWindow()
    window.resize(300, 300)
    window.show()
    sys.exit(app.exec())
````

## วิธีที่ 1: สไตล์ชีตแบบอินไลน์

นำสไตล์ชีตไปใช้โดยตรงกับวิดเจ็ตแต่ละตัวโดยใช้ `setStyleSheet()`

**ข้อดี:**
- ตกแต่งอย่างรวดเร็วสำหรับวิดเจ็ตเฉพาะ

- การตกแต่งกระจัดกระจายทั่วโค้ด
- ยากต่อการดูแลรักษาในแอปพลิเคชันขนาดใหญ่
- ไม่สามารถนำกลับมาใช้ใหม่ได้


### ไวยากรณ์สไตล์ชีตคล้าย CSS

ตกแต่งแอปพลิเคชันของคุณโดยใช้ไวยากรณ์คล้าย CSS ภายในเครื่องหมายอัญประกาศสามตัว

**ไวยากรณ์พื้นฐาน:**
```python
widget.setStyleSheet("""
    background-color: #262626;
    color: #FFFFFF;
    font-family: Arial;
    font-size: 18px;
    border: 1px solid gray;
    border-radius: 5px;
    padding: 5px;
""")
```

**คุณสมบัติทั่วไป:**
- `background-color` - สีพื้นหลังวิดเจ็ต
- `color` - สีข้อความ
- `font-family` - ประเภทฟอนต์ (Arial, Times, ฯลฯ)
- `font-size` - ขนาดฟอนต์เป็นพิกเซล
- `border` - สไตล์ขอบ
- `border-radius` - มุมมน
- `padding` - ระยะห่างภายใน
- `margin` - ระยะห่างภายนอก

**เอกสารอ้างอิงคุณสมบัติทั้งหมด:**
https://doc.qt.io/qt-5/stylesheet-reference.html


````python
import sys
from PySide6.QtCore import Qt
from PySide6.QtWidgets import QApplication, QLabel

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    w = QLabel("This is a placeholder text")
    w.setAlignment(Qt.AlignCenter)
    w.setStyleSheet("""
        background-color: #262626;
        color: #FFFFFF;
        font-family: Titillium;
        font-size: 18px;
        """)
    w.show()
    sys.exit(app.exec())
````

## วิธีที่ 2: ไฟล์ QSS ภายนอก

จัดเก็บสไตล์ชีตในไฟล์ `.qss` ภายนอกและใช้กับทั้งแอปพลิเคชัน

**ข้อดี:**
- การจัดการสไตล์แบบรวมศูนย์
- ง่ายต่อการดูแลรักษาและอัพเดท
- สามารถใช้กับทั้งแอปพลิเคชันพร้อมกัน
- นำกลับมาใช้ใหม่ได้ในหลายโปรเจ็ค
- แยกความกังวล (โค้ดและสไตล์)

**ข้อเสีย:**
- ต้องโหลดไฟล์ตอนรันไทม์
- การตั้งค่าซับซ้อนกว่าเล็กน้อย

**วิธีใช้:**

1. **สร้างไฟล์ `style.qss`:**
```css
QLabel {
    color: black;
    background-color: white;
}

    font-size: 20px;
    font-weight: bold;
}

QPushButton {
    background-color: #2ABf9E;
    color: white;
    padding: 10px;
    border-radius: 5px;
}

QPushButton:hover {
    background-color: #20a085;
}
```

2. **ใช้ในโค้ด Python:**
```python
# สร้างแอปพลิเคชัน

# โหลดและใช้สไตล์ชีต

```

**ประเภท Selector:**

- `QLabel:hover` - ตกแต่งเมื่อชี้เมาส์
- `QLabel:pressed` - ตกแต่งเมื่อกด

ลองสิ่งนี้ด้วยและโดยไม่ตั้งค่า qss เป็นสไตล์ชีตและดูความแตกต่าง


````python
import sys
import re
from PySide6.QtWidgets import (QApplication, QWidget, QVBoxLayout, QHBoxLayout,
                               QLabel, QListWidget, QListWidgetItem, QPushButton)
from PySide6.QtCore import Qt


class Widget(QWidget):
    def __init__(self, parent=None):
        super(Widget, self).__init__(parent)

        menu_widget = QListWidget()
        for i in range(10):
            item = QListWidgetItem(f"Item {i}")
            item.setTextAlignment(Qt.AlignCenter)
            menu_widget.addItem(item)

        text_widget = QLabel("Hello I am here")
        button = QPushButton("Something")

        content_layout = QVBoxLayout()
        content_layout.addWidget(text_widget)
        content_layout.addWidget(button)
        main_widget = QWidget()
        main_widget.setLayout(content_layout)

        layout = QHBoxLayout()
        layout.addWidget(menu_widget, 1)
        layout.addWidget(main_widget, 4)

        self.setLayout(layout)

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    w = Widget()
    w.show()

    with open("style_list.qss", "r") as f:
        _style = f.read()
        app.setStyleSheet(_style)

    sys.exit(app.exec())
````
