## 1. OOP คืออะไร?

Object-Oriented Programming (OOP) เป็นรูปแบบการเขียนโปรแกรมที่ใช้ "Object" และ "Class" เป็นหน่วยพื้นฐาน 
- **Class**: เป็นแบบแผนหรือต้นแบบสำหรับสร้าง Object
- **Object**: เป็นอินสแตนซ์ของ Class ที่มีคุณสมบัติ (Attribute) และพฤติกรรม (Method)

## 2. หลักการสำคัญของ OOP

### 2.1 Encapsulation (การห่อหุ้ม)
การซ่อนรายละเอียดภายใน และให้เข้าถึงผ่าน interface ที่ก่อตั้งไว้เท่านั้น
- Public: เข้าถึงได้จากที่ใดก็ได้
- Private: เข้าถึงได้เฉพาะภายใน Class เท่านั้น (ใช้ underscore `_`)

```python
# ตัวอย่างที่ 1: Class และ Object พื้นฐาน

class Animal:
    """Class สำหรับสัตว์"""
    
    def __init__(self, name, age):
        """Constructor - กำหนดคุณสมบัติเริ่มต้น"""
        self.name = name
        self.age = age
    
    def speak(self):
        """Method - พฤติกรรมของสัตว์"""
        return f"{self.name} is making a sound"

# สร้าง Object จาก Class
dog = Animal("Buddy", 5)
cat = Animal("Whiskers", 3)

print(dog.speak())
print(cat.speak())
```

```python
# ตัวอย่างที่ 2: Encapsulation

class BankAccount:
    """Class บัญชีธนาคาร"""
    
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute (ใช้ __)
    
    def deposit(self, amount):
        """ฝากเงิน"""
        if amount > 0:
            self.__balance += amount
            return f"ฝากเงิน {amount} บาท สำเร็จ"
        return "จำนวนเงินต้องมากกว่า 0"
    
    def withdraw(self, amount):
        """ถอนเงิน"""
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"ถอนเงิน {amount} บาท สำเร็จ"
        return "จำนวนเงินไม่เพียงพอ"
    
    def get_balance(self):
        """ดูยอดเงิน"""
        return f"ยอดเงิน: {self.__balance} บาท"

# ใช้งาน
account = BankAccount("สมชาย", 5000)
print(account.deposit(1000))
print(account.withdraw(2000))
print(account.get_balance())
# print(account.__balance)  # Error! ไม่สามารถเข้าถึง private attribute ได้โดยตรง
```

### 2.2 Inheritance (การสืบทอด)
Class สามารถสืบทอดคุณสมบัติและเมธอดจาก Class อื่นได้ (Parent Class → Child Class)

```python
# ตัวอย่างที่ 3: Inheritance

class Vehicle:
    """Parent Class - พารเนิด"""
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def drive(self):
        return f"{self.brand} {self.model} is driving"

class Car(Vehicle):
    """Child Class สืบทอดจาก Vehicle"""
    
    def __init__(self, brand, model, num_doors):
        super().__init__(brand, model)  # เรียก constructor ของ Parent
        self.num_doors = num_doors
    
    def open_trunk(self):
        return f"Opening trunk of {self.brand} {self.model}"

class Motorcycle(Vehicle):
    """Child Class อีกตัวหนึ่ง"""
    
    def wheelie(self):
        return f"{self.brand} is doing a wheelie!"

# ใช้งาน
car = Car("Toyota", "Camry", 4)
bike = Motorcycle("Honda", "CBR")

print(car.drive())
print(car.open_trunk())
print(bike.wheelie())
```

### 2.3 Polymorphism (การมีหลายรูป)
ความสามารถในการใช้วัตถุหลาย Class ด้วยวิธีเดียวกัน

```python
# ตัวอย่างที่ 4: Polymorphism

class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof! Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Duck(Animal):
    def speak(self):
        return "Quack!"

# ใช้ polymorphism
animals = [Dog(), Cat(), Duck()]

for animal in animals:
    print(animal.speak())  # เรียก speak() ของแต่ละ class แม้ข้างในลูปเดียวกัน
```

### 2.4 Abstraction (การสรุป)
การซ่อนความซับซ้อนและแสดงเฉพาะฟีเจอร์ที่จำเป็น

```python
# ตัวอย่างที่ 5: Abstraction (Abstract Base Class)

from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract Class - คลาสนามธรรม"""
    
    @abstractmethod
    def area(self):
        """เมธอด abstract - ต้องถูก override ใน child class"""
        pass
    
    @abstractmethod
    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2
    
    def perimeter(self):
        return 2 * 3.14 * self.radius

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

# ใช้งาน
circle = Circle(5)
rect = Rectangle(4, 6)

print(f"Circle area: {circle.area()}")
print(f"Rectangle perimeter: {rect.perimeter()}")
```

## 3. โปรเจคจริง: ระบบจัดการห้องสมุด

ตัวอย่างโปรเจคที่รวม OOP 4 หลักการ

```python
# ตัวอย่างที่ 6: Library Management System

from abc import ABC, abstractmethod
from datetime import datetime, timedelta

class Item(ABC):
    """Abstract class สำหรับรายการในห้องสมุด"""
    
    def __init__(self, title, item_id):
        self.title = title
        self.item_id = item_id
        self._is_available = True
    
    @abstractmethod
    def get_info(self):
        pass

class Book(Item):
    """หนังสือ"""
    
    def __init__(self, title, item_id, author, pages):
        super().__init__(title, item_id)
        self.author = author
        self.pages = pages
    
    def get_info(self):
        return f"📚 {self.title} by {self.author} ({self.pages} pages)"

class Magazine(Item):
    """วารสาร"""
    
    def __init__(self, title, item_id, issue):
        super().__init__(title, item_id)
        self.issue = issue
    
    def get_info(self):
        return f"📰 {self.title} - Issue {self.issue}"

class Library:
    """ห้องสมุด"""
    
    def __init__(self, name):
        self.name = name
        self.__items = []  # Private
        self.__borrowers = {}
    
    def add_item(self, item):
        """เพิ่มรายการเข้าห้องสมุด"""
        self.__items.append(item)
        return f"✅ เพิ่ม {item.title} สำเร็จ"
    
    def borrow_item(self, item_id, borrower_name):
        """ยืมรายการจากห้องสมุด"""
        for item in self.__items:
            if item.item_id == item_id:
                if item._is_available:
                    item._is_available = False
                    self.__borrowers[item_id] = {
                        'borrower': borrower_name,
                        'date': datetime.now(),
                        'due_date': datetime.now() + timedelta(days=14)
                    }
                    return f"✅ {borrower_name} ยืม {item.title} สำเร็จ"
                return f"❌ {item.title} ถูกยืมแล้ว"
        return "❌ ไม่พบรายการนี้"
    
    def return_item(self, item_id):
        """คืนรายการ"""
        for item in self.__items:
            if item.item_id == item_id:
                item._is_available = True
                if item_id in self.__borrowers:
                    del self.__borrowers[item_id]
                return f"✅ คืน {item.title} สำเร็จ"
        return "❌ ไม่พบรายการนี้"
    
    def show_available_items(self):
        """แสดงรายการที่ว่าง"""
        print(f"\n📖 {self.name} - รายการที่ว่าง:\n")
        available = [item for item in self.__items if item._is_available]
        for item in available:
            print(f"  {item.get_info()}")
        if not available:
            print("  ไม่มีรายการว่าง")

# ใช้งาน
library = Library("สมุดหลัก")

# เพิ่มรายการ
print(library.add_item(Book("Python for Beginners", "B001", "John Doe", 450)))
print(library.add_item(Book("OOP Mastery", "B002", "Jane Smith", 380)))
print(library.add_item(Magazine("Tech Weekly", "M001", "2024-01")))

# แสดงรายการว่าง
library.show_available_items()

# ยืมหนังสือ
print("\n" + library.borrow_item("B001", "สมชาย"))
print(library.borrow_item("M001", "ศรีนา"))

# แสดงรายการว่าง (หลังยืม)
library.show_available_items()

# คืนหนังสือ
print("\n" + library.return_item("B001"))

# แสดงรายการว่าง (หลังคืน)
library.show_available_items()
```

```python
from abc import ABC, abstractmethod

class Room(ABC):
    """Class นามธรรมสำหรับห้อง"""
    
    def __init__(self, length, width):
        self.length = length
        self.width = width

    @abstractmethod
    def get_purpose(self):
        """กำหนดวัตถุประสงค์ของห้อง"""
        pass

    @abstractmethod
    def get_recommended_lighting(self):
        """ดัชนีแสงที่แนะนำสำหรับห้อง"""
        pass

    def calculate_area(self):
        """คำนวณพื้นที่ห้อง"""
        return self.length * self.width
    
    def describe_room(self):
        """อธิบายรายละเอียดห้อง"""
        area = self.calculate_area()
        return f"A {self.__class__.__name__} of {area} sq ft used for {self.get_purpose()}"

class Bedroom(Room):
    """ห้องนอน"""
    
    def __init__(self, length, width, bed_size):
        super().__init__(length, width)
        self.bed_size = bed_size 

    def get_purpose(self):
        return "sleeping and resting"

    def get_recommended_lighting(self):
        return 20  # lumens per square foot

class Kitchen(Room):
    """ห้องครัว"""
    
    def __init__(self, length, width, has_island=True):
        super().__init__(length, width)
        self.has_island = has_island
    
    def calculate_counter_space(self):
        """คำนวณพื้นที่เคาน์เตอร์"""
        area = self.calculate_area()
        if self.has_island:
            island = area * 1/5
            wall = area * 1/4
        else:
            island = 0
            wall = area * 1/2
        return island, wall
        
    def get_purpose(self):
        return "cooking and food preparation"

    def get_recommended_lighting(self):
        return 70  # lumens per square foot

# ใช้งาน
bedroom = Bedroom(5, 5, 6.3)
print(bedroom.describe_room())
print(f"Recommended lighting: {bedroom.get_recommended_lighting()} lumens/sqft")
print()

kitchen = Kitchen(4, 4, True)
print(kitchen.describe_room())
counter, wall = kitchen.calculate_counter_space()
print(f"Island counter: {counter} sqft, Wall counter: {wall} sqft")
```

```python
import datetime

class Cat:
    # Class Attribute: ตัวแปรที่ใช้นับจำนวนแมว
    total_cats = 0

    def __init__(self, name, age, owner_name):
        self.name = name
        self.age = age
        self.owner_name = owner_name
        
        # กำหนดวันที่เข้ารักษาเป็นเวลาปัจจุบัน
        self.date_in = datetime.datetime.now()
        
        # วันที่ออก (สมมติเริ่มต้นเป็น None)
        self.date_out = None
        
        # นับจำนวนแมวเพิ่มทันที
        Cat.total_cats += 1

    def greet(self):
        """แมวทักทาย"""
        return f"Meow! My name is {self.name}. My owner is {self.owner_name}."

    def print_details(self):
        """แสดงรายละเอียด"""
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Owner: {self.owner_name}")
        print(f"Date In: {self.date_in}")
        print(f"Date Out: {self.date_out}")
        print("-" * 20)
        
    def set_date_out(self, days_stayed):
        """กำหนดวันออก"""
        self.date_out = self.date_in + datetime.timedelta(days=days_stayed)

# ใช้งาน
cat1 = Cat("Leo", 2, "John")
cat2 = Cat("Mimi", 1, "Sarah")

print(cat1.greet())
cat2.set_date_out(7)  # พัก 7 วัน

cat1.print_details()
cat2.print_details()

print(f"Total cats: {Cat.total_cats}")
```

```python
class BankAccount:
    # Class Attributes: ตัวแปรที่ใช้ร่วมกันทั้งคลาส
    branch_name = "KKU Complex"
    branch_number = 1724
    last_loan_number = 0
    last_saving_number = 0

    __type_saving = 1
    __type_loan = 2
    
    def __init__(self, name, account_type="saving", initial_amount=0):
        self.name = name
        self.type = account_type
        self.balance = initial_amount

        # สร้างหมายเลขบัญชีอัตโนมัติตามประเภท
        if self.type == "saving":
            type_code = BankAccount.__type_saving
            BankAccount.last_saving_number += 1
            running_number = BankAccount.last_saving_number
        else:
            type_code = BankAccount.__type_loan
            BankAccount.last_loan_number += 1
            running_number = BankAccount.last_loan_number

        self.account_number = f"{BankAccount.branch_number}-{type_code}-{running_number}"

    def print_customer(self):
        print("----- Customer Record -----")
        print(f"Name: {self.name}")
        print(f"Account number: {self.account_number}")
        print(f"Account type: {self.type}")
        print(f"Balance: {self.balance}")
        print("----- End Record -----")
    
    def deposit(self, amount=0):
        """ฝากเงิน"""
        self.balance += amount
        return self.balance
    
    def pay_loan(self, amount=0):
        """ชำระเงินกู้"""
        self.balance += amount
        return self.balance

# ใช้งาน
john = BankAccount("John", "saving", 500)
tim = BankAccount("Tim", "loan", -1000000)

john.deposit(3000)
tim.pay_loan(500000)

john.print_customer()
tim.print_customer()
```
