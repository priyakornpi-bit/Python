# 📌 MODULE 1: บทนำ OOP (Introduction)

## 🤔 OOP คืออะไร?
**Object-Oriented Programming (OOP)** คือ "กระบวนทัศน์" (Paradigm) การเขียนโปรแกรม ที่จำลองโลกแห่งความเป็นจริงให้อยู่ในรูปของ **วัตถุ (Objects)** ที่มีข้อมูลและพฤติกรรมของตัวเอง

## 🎯 ทำไมต้องเรียน OOP?
- ❌ **ปัญหาแบบ Procedural:** โค้ดยาว, ซ้ำ ๆ, ยากต่อบำรุงรักษา
- ✅ **วิธีแก้ด้วย OOP:** โครงสร้างชัดเจน, นำกลับมาใช้ได้, ขยายได้ง่าย

## 📍 ใช้เมื่อไร?
- โครงการขนาดกลางถึงใหญ่
- ต้องการจัดกลุ่มข้อมูลและฟังก์ชันที่เกี่ยวข้องไว้ด้วยกัน
- ต้องการสร้าง "แม่แบบ" เพื่อสร้างสิ่งของหลายตัว

## ⚠️ ถ้าไม่ใช้ OOP จะเกิดอะไร?
- โค้ด Procedural ที่ทำซ้ำ ๆ
- ยากต่อการแก้ไข (หลายที่ต้องแก้)
- ยากต่อการเพิ่มฟีเจอร์ใหม่
- "Spaghetti code" - วุ่นวายและไม่มีโครงสร้าง

### 💡 Insight: Procedural vs OOP

````python
# ❌ Procedural (ห้ามม!)
dog1_name = "Buddy"
dog1_breed = "Golden"
dog1_age = 5

dog2_name = "Lucy"
dog2_breed = "Labrador"
dog2_age = 2

# ต้องทำอะไรสักอย่างซ้ำ ๆ สำหรับแต่ละสุนัข...
print(f"Procedural: {dog1_name} is a {dog1_breed}")
print(f"Procedural: {dog2_name} is a {dog2_breed}")

# ✅ OOP (ดีมาก!)
class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age
    
    def describe(self):
        return f"{self.name} is a {self.breed}"

# สร้างอ็อบเจกต์
dog1 = Dog("Buddy", "Golden", 5)
dog2 = Dog("Lucy", "Labrador", 2)

# เรียกใช้อย่างเดียวกัน (ไม่ต้องซ้ำโค้ด)
print(f"OOP: {dog1.describe()}")
print(f"OOP: {dog2.describe()}")
````

# 🏗️ MODULE 2: Classes & Objects

## 📚 คลาส (Class) vs อ็อบเจกต์ (Object)

### 🎨 **คลาส = พิมพ์เขียว (Blueprint)**
- เป็น "แม่แบบ" ที่กำหนด **โครงสร้าง** ของข้อมูลและพฤติกรรม
- ตัวอย่าง: เครื่องตัดคุกกี้ = คลาส
- ไม่มีตัวตนจริง

### 🎁 **อ็อบเจกต์ = ตัวจริง (Instance)**
- เป็น "สิ่งของจริง" ที่สร้างตามแม่แบบ
- ตัวอย่าง: คุกกี้ที่เพิ่งทำ = อ็อบเจกต์
- มี **สถานะ (State)** เป็นของตัวเอง

## 🔗 ความสัมพันธ์
```
Class Dog          →  dog1, dog2, dog3 (Objects/Instances)
พิมพ์เขียว  →  สิ่งของจริง
```

### 📚 ศัพท์เทคนิคที่ต้องรู้

| ศัพท์ | ความหมาย | ตัวอย่าง |
|------|----------|----------|
| **Base Class** | คลาสทั่วไปที่อยู่ด้านบน | `Vehicle` |
| **Superclass / Parent** | คลาสแม่ที่มีคุณสมบัติทั่วไป | `Animal` |
| **Subclass / Child** | คลาสลูกที่สืบทอด + เพิ่มเติม | `Dog(Animal)` |
| **IS-A Relationship** | ความสัมพันธ์ "เป็น" | สุนัข IS-A สัตว์ |
| **Object Instance** | ของจริงที่สร้างจากคลาส | `my_dog = Dog()` |

````python
# ก่อนอื่น กำหนดคลาส (พิมพ์เขียว)
class Dog:
    """พิมพ์เขียวสำหรับสร้างสุนัข"""
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age
    
    def bark(self):
        return f"{self.name} says Woof! 🐕"
    
    def describe(self):
        return f"{self.name} is a {self.age}-year-old {self.breed}"

print("✅ คลาส Dog ถูกกำหนดแล้ว")
````

````python
# สร้างอ็อบเจกต์ (instances) จากคลาส
my_dog = Dog("Buddy", "Golden Retriever", 5)
your_dog = Dog("Lucy", "Labrador", 2)

print(f"สุนัขของฉัน: {my_dog.name} ({my_dog.breed})")
print(f"สุนัขของเธอ: {your_dog.name} ({your_dog.breed})")
print()
print(my_dog.bark())
print(your_dog.bark())
print()
print(my_dog.describe())
print(your_dog.describe())
````

### 💡 Insight: Object Identity

แต่ละ object มี **ID ที่ไม่ซ้ำกัน** ในหน่วยความจำ ถึงแม้ค่า attributes จะเหมือนกัน

````python
# การเปรียบเทียบ: == vs is
dog_a = Dog("Rex", "Poodle", 3)
dog_b = Dog("Rex", "Poodle", 3)  # ข้อมูลเหมือน แต่เป็น object ต่างตัว

print(f"dog_a == dog_b: {dog_a == dog_b}")  # False (object ต่างตัว)
print(f"dog_a is dog_b: {dog_a is dog_b}")  # False
print(f"id(dog_a): {id(dog_a)}")
print(f"id(dog_b): {id(dog_b)}")
print()

# การเปรียบเทียบ: object เดียวตัว
dog_c = dog_a
print(f"dog_a is dog_c: {dog_a is dog_c}")  # True (object เดียวตัว)
````

### 🚨 Pitfall: ผิดพลาดที่พบบ่อย

**❌ ผิด:** "class และ object เป็นสิ่งเดียวกัน"

**✅ ถูก:** Class คือ "แม่แบบ" / Object คือ "สิ่งของจริง"

# 🎪 MODULE 3: Attributes & Methods

## 🔍 Attributes (แอตทริบิวต์) = ข้อมูล
- **ตัวแปร** ของ object
- เก็บ "สถานะ" (state) ของ object
- เข้าถึงด้วย: `object.attribute`

## 🔧 Methods (เมธอด) = พฤติกรรม
- **ฟังก์ชัน** ของ object
- ทำงานกับข้อมูล (attributes)
- เรียกใช้ด้วย: `object.method()`

## 📍 พารามิเตอร์ `self`
- **ตัวแทนของ object ที่กำลังทำงาน**
- ใช้ `self.attribute` เพื่อเข้าถึงข้อมูลของ object นั้น

````python
class Student:
    """แสดง Instance Attributes และ __init__ Constructor"""
    
    def __init__(self, name, student_id, gpa):
        # Instance Attributes (เป็นของแต่ละ object)
        self.name = name
        self.student_id = student_id
        self.gpa = gpa
    
    def get_info(self):
        """เมธอด: ส่งคืนข้อมูล"""
        return f"{self.name} (ID: {self.student_id}) - GPA: {self.gpa}"
    
    def study(self, hours):
        """เมธอด: ศึกษาเป็นเวลา N ชั่วโมง"""
        return f"{self.name} studied for {hours} hours and improved GPA"

# สร้าง 2 instance
s1 = Student("Alice", "6501001", 3.8)
s2 = Student("Bob", "6501002", 3.5)

print(s1.get_info())
print(s2.get_info())
print()
print(s1.study(3))
````

### 📊 Constructor (`__init__`)

- **เมธอดพิเศษ** ที่ถูกเรียกอัตโนมัติเมื่อสร้าง object
- ใช้กำหนดค่าเริ่มต้นให้ attributes
- ชื่อต้องเป็น `__init__` (มีขีดล่าง 2 อันหน้าหลัง)

### 💡 Insight: Class vs Instance Attributes

````python
class Car:
    # Class Attribute (ใช้ร่วมกันทุก objects)
    wheels = 4
    
    def __init__(self, brand, color):
        # Instance Attributes (เป็นของแต่ละ object)
        self.brand = brand
        self.color = color

car1 = Car("Toyota", "Red")
car2 = Car("Honda", "Blue")

print(f"Car1: {car1.brand}, {car1.color}, wheels: {car1.wheels}")
print(f"Car2: {car2.brand}, {car2.color}, wheels: {car2.wheels}")
print()

# เปลี่ยน instance attribute (เฉพาะ car1)
car1.color = "Green"
print(f"Car1 color เปลี่ยนเป็น: {car1.color}")
print(f"Car2 color ยังคงเป็น: {car2.color}")
````

### 🚨 Pitfall: Class Attribute vs Instance Attribute

````python
class Counter:
    count = 0  # Class Attribute
    
    def __init__(self):
        self.value = 0  # Instance Attribute

obj1 = Counter()
obj2 = Counter()

# ❌ ผิด: เปลี่ยน class attribute ผ่าน instance
obj1.count = 10  # สร้าง instance attribute ใหม่ขึ้นมา (ไม่ได้เปลี่ยน class attribute)

print(f"obj1.count: {obj1.count}")  # 10 (instance attribute)
print(f"obj2.count: {obj2.count}")  # 0 (ยังใช้ class attribute)
print(f"Counter.count: {Counter.count}")  # 0 (class attribute ยังเหมือนเดิม)
````

# 🔐 MODULE 4: Encapsulation (การห่อหุ้ม)

## 🤔 ทำไมต้องเก็บเสน่ห์ข้อมูล?
- ป้องกัน object ไม่ให้มีสถานะ "ผิด"
- ควบคุมวิธีเข้าถึงข้อมูล
- เพิ่มความปลอดภัย (validation)

## 🔒 Private Attributes
ใน Python ใช้เครื่องหมาย `__` (double underscore) นำหน้า
- `self.__balance` → "private" (ไม่ควรเข้าถึงจากภายนอก)
- `self._balance` → "protected" (อย่าแตะหน่อย)

## 📊 Getter/Setter
ใช้เมธอดเพื่อเข้าถึงข้อมูลแบบควบคุมได้

````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # Private attribute
    
    def get_balance(self):
        """Getter: เข้าถึงยอดเงิน"""
        return self.__balance
    
    def deposit(self, amount):
        """ฝากเงิน (พร้อม validation)"""
        if amount > 0:
            self.__balance += amount
            return f"✅ ฝากเงิน ${amount} สำเร็จ ยอดเงินใหม่: ${self.__balance}"
        return "❌ จำนวนเงินต้องเป็นบวก"
    
    def withdraw(self, amount):
        """ถอนเงิน (พร้อม validation)"""
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"✅ ถอนเงิน ${amount} สำเร็จ ยอดเงินใหม่: ${self.__balance}"
        return "❌ จำนวนเงินไม่ถูกต้องหรือไม่พอ"

# ใช้งาน
acc = BankAccount("Alice", 1000)
print(f"ยอดเงินเริ่มต้น: ${acc.get_balance()}")
print(acc.deposit(500))
print(acc.withdraw(200))
print(acc.withdraw(2000))  # ถอนเกินยอด
````

### 📌 @property Decorator (วิธี Pythonic)

````python
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius
    
    @property
    def celsius(self):
        """ได้ค่า Celsius"""
        return self.__celsius
    
    @celsius.setter
    def celsius(self, value):
        """ตั้งค่า Celsius (พร้อม validation)"""
        if value < -273.15:
            print("❌ อุณหภูมิต่ำเกินไป!")
        else:
            self.__celsius = value
    
    @property
    def fahrenheit(self):
        return (self.__celsius * 9/5) + 32

temp = Temperature(25)
print(f"Celsius: {temp.celsius}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")
print()

# ตั้งค่าแบบเหมือนแอตทริบิวต์ (เบื้องหลังเรียก setter)
temp.celsius = 30
print(f"อุณหภูมิใหม่: {temp.celsius}°C = {temp.fahrenheit}°F")
````

# 👨‍👩‍👧 MODULE 5: Inheritance (การสืบทอด)

## 🧬 Inheritance คืออะไร?
คลาสลูก **สืบทอด** คุณสมบัติและเมธอดจากคลาสแม่

## 🔗 IS-A Relationship
```
Dog IS-A Animal
Cat IS-A Animal
```

## 📍 ชนิด
- **Base Class (Parent):** คลาสที่ถูกสืบทอด
- **Derived Class (Child):** คลาสที่สืบทอด

````python
# คลาสแม่
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound"

# คลาสลูก
class Dog(Animal):  # สืบทอดจาก Animal
    def __init__(self, name, breed):
        super().__init__(name)  # เรียก parent constructor
        self.breed = breed
    
    def speak(self):  # Override parent method
        return f"{self.name} says Woof! 🐕"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
    
    def speak(self):
        return f"{self.name} says Meow! 🐱"

# ใช้งาน
dog = Dog("Buddy", "Golden")
cat = Cat("Whiskers", "Orange")

print(dog.speak())
print(cat.speak())
````

### 📊 `super()` Function

````python
class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def info(self):
        return f"{self.brand} ({self.year})"

class Car(Vehicle):
    def __init__(self, brand, year, doors):
        super().__init__(brand, year)  # เรียก parent __init__
        self.doors = doors
    
    def info(self):
        parent_info = super().info()  # เรียก parent method
        return f"{parent_info} - {self.doors} doors"

car = Car("Toyota", 2023, 4)
print(car.info())
````

### 💡 Insight: Multi-level Inheritance

````python
class LivingBeing:
    def __init__(self, name):
        self.name = name
    
    def alive(self):
        return f"{self.name} is alive"

class Animal(LivingBeing):
    def __init__(self, name, species):
        super().__init__(name)
        self.species = species

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name, "Canine")
        self.breed = breed

dog = Dog("Buddy", "Golden")
print(dog.alive())  # ได้จาก LivingBeing
print(f"Species: {dog.species}")  # ได้จาก Animal
print(f"Breed: {dog.breed}")  # ได้จาก Dog
````

### 🚨 Pitfall: ลืม `super().__init__()`

````python
# ❌ ผิด: ลืม super()
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        # ❌ ไม่ได้เรียก super().__init__(name)
        self.age = age

child = Child("Alice", 10)
print(f"Age: {child.age}")  # ✅ OK
# print(f"Name: {child.name}")  # ❌ AttributeError: 'Child' object has no attribute 'name'
````

````python
# Polymorphism ผ่าน Inheritance
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof! 🐕"

class Cat(Animal):
    def speak(self):
        return "Meow! 🐱"

class Duck(Animal):
    def speak(self):
        return "Quack! 🦆"

# Polymorphism in Action
animals = [Dog(), Cat(), Duck()]

for animal in animals:
    print(animal.speak())  # เมธอด speak() เดียวกัน แต่ผลลัพธ์ต่างกัน
````

### 📊 Operator Overloading

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """Override + operator"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        """Override string representation"""
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2  # ใช้ __add__

print(f"{v1} + {v2} = {v3}")
````

````python
### 📊 ตัวอย่าง Multi-level Inheritance (สืบทอด 3 ชั้น)

class Vehicle:
    """ยานพาหนะ (คลาสแม่ที่สุด)"""
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    
    def start_engine(self):
        return f"{self.brand} {self.model}: เครื่องยนต์สตาร์ทแล้ว"

class Car(Vehicle):
    """รถยนต์ (ชั้นที่ 2)"""
    def __init__(self, brand, model, num_wheels=4):
        super().__init__(brand, model)
        self.num_wheels = num_wheels
    
    def drive(self):
        return f"รถยนต์ขับเคลื่อนบน {self.num_wheels} ล้อ"

class ElectricCar(Car):
    """รถยนต์ไฟฟ้า (ชั้นที่ 3)"""
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
    
    def charge(self):
        return f"กำลังชาร์จแบตเตอรี่ {self.battery_capacity} kWh"
    
    def start_engine(self):  # Override - เปลี่ยนพฤติกรรมเพราะเป็นไฟฟ้า
        return f"{self.brand} {self.model}: มอเตอร์ไฟฟ้าเปิดแล้ว"

# ทดสอบ
ecar = ElectricCar("Tesla", "Model S", 40000)
print(ecar.start_engine())  # ใช้ของ ElectricCar
print(ecar.drive())         # ใช้ของ Car (สืบทอดมา)
print(ecar.charge())        # ใช้ของ ElectricCar เอง
````

### 🔄 Method Overriding (เขียนทับเมธอด)

**ความหมาย:** คลาสลูกสร้าง version ใหม่ของเมธอดที่สืบทอดมา

**เมื่อใช้:**
- เมื่อคลาสแม่ "เกือบ" ตรงกับที่ต้องการ
- ต้องการเฉพาะบางเมธอดที่ต่างออกไป
- ตัวอย่าง: ระบบการชำระเงิน (Credit Card, PayPal, PromptPay ต่างกัน)

````python
### 💳 ตัวอย่าง: ระบบการชำระเงินหลากหลายวิธี

class Payment:
    """คลาสแม่: วิธีการชำระเงินทั่วไป"""
    def __init__(self, amount):
        self.amount = amount
    
    def validate(self):
        """ตรวจสอบจำนวนเงิน"""
        return self.amount > 0
    
    def process(self):
        print(f"ชำระเงิน ${self.amount} แบบทั่วไป")

class CreditCardPayment(Payment):
    """เมธอด Overriding: ชำระด้วยบัตรเครดิต"""
    def __init__(self, amount, card_number):
        super().__init__(amount)
        self.card_number = card_number
    
    def validate(self):  # ← Override: ตรวจสอบเพิ่มเติม
        return super().validate() and len(str(self.card_number)) == 16
    
    def process(self):  # ← Override: ฝากด้วยวิธีต่างกัน
        if self.validate():
            print(f"✅ ชำระ ${self.amount} ด้วยบัตร {self.card_number}")
        else:
            print("❌ บัตรไม่ถูกต้อง")

class PayPalPayment(Payment):
    """เมธอด Overriding: ชำระด้วย PayPal"""
    def __init__(self, amount, email):
        super().__init__(amount)
        self.email = email
    
    def validate(self):  # ← Override
        return super().validate() and '@' in self.email
    
    def process(self):  # ← Override
        if self.validate():
            print(f"✅ ชำระ ${self.amount} ผ่าน PayPal ({self.email})")
        else:
            print("❌ อีเมลไม่ถูกต้อง")

class PromptPayPayment(Payment):
    """เมธอด Overriding: ชำระด้วย PromptPay"""
    def __init__(self, amount, phone_number):
        super().__init__(amount)
        self.phone_number = phone_number
    
    def validate(self):  # ← Override
        return (super().validate() and 
                len(self.phone_number) == 10 and 
                self.phone_number[0] == "0")
    
    def process(self):  # ← Override
        if self.validate():
            print(f"✅ ชำระ ${self.amount} ผ่าน PromptPay ({self.phone_number})")
        else:
            print("❌ หมายเลขโทรศัพท์ไม่ถูกต้อง")

# ทดสอบ
print("=== ทดสอบการชำระเงิน ===")
cc = CreditCardPayment(1000, 4900440012345678)
cc.process()

pp = PayPalPayment(2000, "user@gmail.com")
pp.process()

prompt = PromptPayPayment(500, "0921112222")
prompt.process()
````

### 👯 Multiple Inheritance (สืบทอดจากหลายคลาส)

**ความหมาย:** คลาสลูกมีพ่อแม่หลายคน

**ไวยากรณ์:**
```python
class Child(Parent1, Parent2, Parent3):
    pass
```

**สำคัญ:** Python ใช้ **Method Resolution Order (MRO)** เพื่อตัดสินใจเมื่อเมธอดมีชื่อเดียวกัน
- ตรวจสอบด้วย: `ClassName.__mro__`
- ลำดับปกติ: คลาสตัวเอง → Parent1 → Parent2 → ... → object

````python
### 📱 ตัวอย่าง: SmartPhone (สืบทอดจาก 3 คลาส)

# คลาสแม่ที่ 1: โทรศัพท์
class Phone:
    def __init__(self, phone_number):
        self.phone_number = phone_number
    
    def call(self, number):
        return f"📞 โทรหา {number} จากเบอร์ {self.phone_number}"

# คลาสแม่ที่ 2: กล้อง
class Camera:
    def __init__(self, megapixels):
        self.megapixels = megapixels
    
    def take_photo(self):
        return f"📷 ถ่ายรูป {self.megapixels}MP"

# คลาสแม่ที่ 3: GPS
class GPSDevice:
    def __init__(self, coordinates):
        self.coordinates = coordinates
    
    def navigate(self, destination):
        return f"🗺️ นำทาง จาก {self.coordinates} ไป {destination}"

# คลาสลูก: SmartPhone (สืบทอดจากทั้ง 3 คลาส)
class SmartPhone(Phone, Camera, GPSDevice):
    def __init__(self, phone_number, megapixels, coordinates, model):
        Phone.__init__(self, phone_number)
        Camera.__init__(self, megapixels)
        GPSDevice.__init__(self, coordinates)
        self.model = model

# ทดสอบ
iphone = SmartPhone("081-234-5678", 48, "13.7563°N, 100.5018°E", "iPhone 15")

print(iphone.call("999"))
print(iphone.take_photo())
print(iphone.navigate("Central World"))

# ตรวจสอบ MRO (Method Resolution Order)
print("\n=== Method Resolution Order ===")
print(SmartPhone.__mro__)
````

### 🎯 Mixins (ผสมผสานคุณสมบัติ)

**ความหมาย:** คลาสที่ให้ functionality เพิ่มเติม **โดยไม่ใช่คลาสแม่จริง**

**ความแตกต่าง:**
- **Inheritance (IS-A):** `Dog IS-A Animal` → มี parent/child relationship
- **Mixin (CAN-DO):** `CollegeStudent CAN-DO GradeCalculator` → เพิ่มคุณสมบัติ

**ข้อดี:**
- หลีกเลี่ยง deep inheritance hierarchies
- โค้ดโมดูล (modular)
- นำกลับใช้ได้ง่าย (reusable)

````python
### 🎨 ตัวอย่าง: Mixins สำหรับนักเรียน

# Mixin 1: คำนวณเกรด
class GradeCalculatorMixin:
    def calculate_average(self, scores):
        return sum(scores) / len(scores) if scores else 0
    
    def get_letter_grade(self, average):
        if average >= 90: return 'A'
        if average >= 80: return 'B'
        if average >= 70: return 'C'
        if average >= 60: return 'D'
        return 'F'

# Mixin 2: ติดตามการเข้าเรียน
class AttendanceMixin:
    def check_attendance(self, days_present, total_days):
        return (days_present / total_days) * 100 if total_days > 0 else 0

# Mixin 3: Logging (บันทึกข้อมูล)
class LoggerMixin:
    def log(self, message):
        print(f"[LOG] {message}")

# คลาสหลัก: นักเรียน
class Student:
    def __init__(self, name, student_id):
        self.name = name
        self.student_id = student_id

# คลาสลูก: นักเรียนระดับมหาวิทยาลัย (ใช้ Mixins)
class CollegeStudent(Student, GradeCalculatorMixin, AttendanceMixin, LoggerMixin):
    pass

# ทดสอบ
student = CollegeStudent("อลิส", "6501001")

# ใช้ GradeCalculatorMixin
grades = [85, 90, 88, 92]
avg = student.calculate_average(grades)
letter = student.get_letter_grade(avg)
print(f"📚 นักเรียน: {student.name}")
print(f"   คะแนนเฉลี่ย: {avg:.1f} → เกรด {letter}")

# ใช้ AttendanceMixin
attendance = student.check_attendance(18, 20)
print(f"   การเข้าเรียน: {attendance}%")

# ใช้ LoggerMixin
student.log(f"{student.name} บันทึกเสร็จสิ้น")
````

### 🦆 Duck Typing

**ความหมาย:** "ถ้ามันเดินเหมือนเป็ด ร้องเสียงเหมือนเป็ด → มันคือเป็ด"

**หมายถึง:** Python ไม่สนใจ **ประเภท** ตราบใดที่มี **เมธอดที่จำเป็น**

**ข้อดี:**
- ไม่ต้องกังวล class hierarchy
- Flexible มาก ๆ
- โค้ดเป็นสากล (generic)

````python
### 🎸 ตัวอย่าง: Duck Typing (หลายคลาสเล่น() ได้)

# คลาส 1
class Guitar:
    def play(self):
        return "🎸 Strumming guitar..."

# คลาส 2
class Piano:
    def play(self):
        return "🎹 Playing piano..."

# คลาส 3 (ไม่สืบทอด Guitar หรือ Piano!)
class Violin:
    def play(self):
        return "🎻 Playing violin..."

# ฟังก์ชัน: เล่นเครื่องดนตรีใดก็ได้ (ตราบใดที่มี play())
def play_music(instrument):
    print(instrument.play())

# ทดสอบ: Duck Typing ทำให้เล่นได้ทั้งหมด!
print("=== Duck Typing ===")
play_music(Guitar())     # ✅ ทำงาน
play_music(Piano())      # ✅ ทำงาน
play_music(Violin())     # ✅ ทำงาน (แม้ไม่สืบทอด)
````

## 🏛️ MODULE 7 (เพิ่มเติม): Abstract Classes (คลาสนามธรรม) & DocStrings

### 📋 Abstract Class

**ความหมาย:**
- คลาสที่มี **ที่น้อยหนึ่งเมธอดแบบนามธรรม** (ไม่มีการ implement)
- **ไม่สามารถสร้าง object ได้โดยตรง**
- ใช้เป็น "สัญญา" (contract) ให้คลาสลูก

**เมื่อใช้:**
- ต้องการบังคับให้คลาสลูก implement เมธอดบางตัว
- ต้องการสร้าง interface (มาตรฐาน)
- ต้องการ "ข้อตกลง" ให้ subclasses ทั้งหมด

**ไวยากรณ์:**
```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):  # ← นามธรรม (ไม่มี implementation)
        pass
```

````python
### 🎨 ตัวอย่าง: Shape (รูปทรง) - Abstract Class

from abc import ABC, abstractmethod
import math

# คลาสนามธรรม
class Shape(ABC):
    """รูปทรงนามธรรม - ต้องมีพื้นที่และเส้นรอบรูป"""
    
    def __init__(self, color="black"):
        self.color = color
    
    @abstractmethod  # ← บังคับให้ subclass implement
    def calculate_area(self):
        pass
    
    @abstractmethod  # ← บังคับให้ subclass implement
    def calculate_perimeter(self):
        pass

# คลาสลูก 1: สี่เหลี่ยมผืนผ้า
class Rectangle(Shape):
    def __init__(self, width, height, color="black"):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def calculate_area(self):  # ← implement abstract method
        return self.width * self.height
    
    def calculate_perimeter(self):  # ← implement abstract method
        return 2 * (self.width + self.height)

# คลาสลูก 2: วงกลม
class Circle(Shape):
    def __init__(self, radius, color="black"):
        super().__init__(color)
        self.radius = radius
    
    def calculate_area(self):  # ← implement abstract method
        return math.pi * self.radius ** 2
    
    def calculate_perimeter(self):  # ← implement abstract method
        return 2 * math.pi * self.radius

# ทดสอบ
print("=== Abstract Classes ===")

# ❌ ไม่สามารถสร้าง object จาก Shape โดยตรง
# shape = Shape()  # TypeError: Can't instantiate abstract class

# ✅ สามารถสร้าง object จาก Rectangle
rect = Rectangle(5, 3)
print(f"📦 สี่เหลี่ยม {rect.width}×{rect.height}")
print(f"   พื้นที่: {rect.calculate_area()}")
print(f"   เส้นรอบรูป: {rect.calculate_perimeter()}")

# ✅ สามารถสร้าง object จาก Circle
circle = Circle(4)
print(f"\n⭕ วงกลม รัศมี {circle.radius}")
print(f"   พื้นที่: {circle.calculate_area():.2f}")
print(f"   เส้นรอบรูป: {circle.calculate_perimeter():.2f}")
````

### 📝 DocStrings (เอกสารประกอบ)

**ความหมาย:** ข้อความอธิบายคลาส/เมธอด ที่เป็นส่วนของโค้ด

**ประโยชน์:**
- อธิบาย **ว่าทำอะไร**
- อธิบาย **พารามิเตอร์** (Args)
- อธิบาย **ค่าที่ส่งกลับ** (Returns)
- อธิบาย **ข้อผิดพลาด** (Raises)
- ให้ **ตัวอย่างใช้งาน**

**ไวยากรณ์:** ใช้ `"""` หรือ `'''` ทั้ง 3 เครื่องหมาย

**วิธีดู:** `help(function)` หรือ `function.__doc__`

````python
### 📖 ตัวอย่าง: DocStrings แบบ Google Style

def calculate_discount(price, discount_rate):
    """คำนวณราคาหลังลดราคา
    
    Args:
        price (float): ราคาเดิม
        discount_rate (float): ลดราคา 0-100%
    
    Returns:
        float: ราคาหลังลดแล้ว
    
    Raises:
        ValueError: ถ้า discount_rate ไม่อยู่ระหว่าง 0-100
    
    Examples:
        >>> calculate_discount(1000, 20)
        800.0
    """
    if not 0 <= discount_rate <= 100:
        raise ValueError("ลดราคาต้องอยู่ระหว่าง 0-100")
    return price * (1 - discount_rate/100)

# ทดสอบ
print("=== DocStrings ===")
print(calculate_discount(1000, 20))

# ดูเอกสาร
print("\n" + "="*50)
print("เอกสารประกอบ:")
print("="*50)
print(calculate_discount.__doc__)
````

````python
### 🎮 ตัวอย่างโปรเจกต์: Game Character System

from abc import ABC, abstractmethod

# Abstract Base Class
class GameEntity(ABC):
    """หน่วยเกมพื้นฐาน"""
    def __init__(self, name):
        self.name = name
        self._alive = True  # weak private
        self.__health_multiplier = 1.0  # truly private
    
    def take_damage(self, damage):
        damage = damage * self.__health_multiplier
        print(f"💔 {self.name} รับดาเมจ {damage}!")
    
    def _check_status(self):  # protected method
        return "🟢 Active" if self._alive else "⚫ Inactive"

# Character Class (Inheritance + Encapsulation)
class Character(GameEntity):
    """ตัวละครท่องเที่ยว"""
    def __init__(self, name, health, level):
        super().__init__(name)
        self.health = health
        self.level = level
        self._experience = 0  # weak private
        self.__skill_points = level * 2  # truly private
    
    def gain_experience(self, amount):
        self._experience += amount
        print(f"⭐ {self.name} ได้ experience {amount}")
        if self._experience >= 100:
            self.__level_up()
    
    def __level_up(self):  # private method
        self.level += 1
        self.__skill_points += 2
        print(f"🎉 {self.name} ขึ้นระดับเป็น {self.level}!")

# Combat Class
class Warrior(Character):
    """นักรบสุดแกร่ง"""
    def __init__(self, name, health, level, weapon):
        super().__init__(name, health, level)
        self.weapon = weapon
    
    def attack(self):
        return f"⚔️  {self.name} ใช้ {self.weapon} โจมตี!"

class Mage(Character):
    """พ่อมดสูงสุด"""
    def __init__(self, name, health, level, magic_type):
        super().__init__(name, health, level)
        self.magic_type = magic_type
    
    def cast_spell(self):
        return f"✨ {self.name} ปล่อยเวทย์ {self.magic_type}!"

# ทดสอบ
print("=== Game Character System ===\n")

warrior = Warrior("Conan", 100, 5, "ดาบยักษ์")
mage = Mage("Merlin", 60, 8, "Fireball")

print(f"⚔️  {warrior.name} - Level {warrior.level}")
print(warrior.attack())
warrior.gain_experience(50)

print(f"\n🧙 {mage.name} - Level {mage.level}")
print(mage.cast_spell())
mage.gain_experience(120)  # ต้องขึ้นระดับ
````

# 🏛️ MODULE 7: Abstract Classes & DocStrings

## 📋 Abstract Classes
- คลาสที่ **ไม่สามารถสร้าง object ได้โดยตรง**
- ใช้เป็น **Interface** สำหรับคลาสลูก
- บังคับให้คลาสลูก implement abstract methods

## 📝 DocStrings
- เอกสารประกอบโค้ดที่เข้าถึงได้แบบโปรแกรม
- ใช้ `"""..."""`
- เรียกดูด้วย `help()` หรือ `.__doc__`

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract base class for all shapes.
    
    Attributes:
        None (abstract)
    
    Methods:
        area(): Calculate area
        perimeter(): Calculate perimeter
    """
    
    @abstractmethod
    def area(self):
        """Calculate the area of the shape."""
        pass
    
    @abstractmethod
    def perimeter(self):
        """Calculate the perimeter of the shape."""
        pass

class Rectangle(Shape):
    """A rectangle shape with width and height."""
    
    def __init__(self, width, height):
        """Initialize rectangle.
        
        Args:
            width (float): Width of rectangle
            height (float): Height of rectangle
        """
        self.width = width
        self.height = height
    
    def area(self):
        """Return the area of rectangle."""
        return self.width * self.height
    
    def perimeter(self):
        """Return the perimeter of rectangle."""
        return 2 * (self.width + self.height)

class Circle(Shape):
    """A circle shape with radius."""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        """Return the area of circle."""
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        """Return the circumference of circle."""
        import math
        return 2 * math.pi * self.radius

# ใช้งาน
rect = Rectangle(5, 3)
circle = Circle(2)

print(f"Rectangle area: {rect.area()}, perimeter: {rect.perimeter()}")
print(f"Circle area: {circle.area():.2f}, circumference: {circle.perimeter():.2f}")
````

### 📖 DocStrings และ `help()`

````python
# ดูเอกสารประกอบ
print(Shape.__doc__)
print()
print(Rectangle.area.__doc__)
print()
help(Rectangle)  # แสดงข้อมูลทั้งหมด
````

# 🎁 MODULE 8: Inner Classes (คลาสภายใน)

## 🔑 Inner Class
- คลาสที่ถูกกำหนดอยู่ **ภายในคลาสอื่น**
- ความสัมพันธ์ **HAS-A** ("มี")
- ทำให้โค้ดมีการจัดระเบียบที่ดีขึ้น

````python
class Car:
    """A car with an engine."""
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.engine = self.Engine()  # สร้าง inner object
    
    class Engine:
        """Inner class representing car engine."""
        
        def __init__(self):
            self.status = "Off"
        
        def start(self):
            self.status = "On"
            return "Engine started! 🚗"
        
        def stop(self):
            self.status = "Off"
            return "Engine stopped!"
    
    def drive(self):
        if self.engine.status == "On":
            return f"{self.brand} {self.model} is driving!"
        return "Start engine first!"

# ใช้งาน
my_car = Car("Toyota", "Corolla")
print(my_car.drive())  # ❌ Engine off
print(my_car.engine.start())  # ✅ Start engine
print(my_car.drive())  # ✅ Now driving
````

## 🎯 ตัวอย่างประยุกต์: Zoo Management System

````python
from abc import ABC, abstractmethod
from datetime import datetime

# 1. Abstract Base Class
class Animal(ABC):
    """Base class for all animals in the zoo."""
    
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
        self.__health = 100  # Private attribute
    
    @abstractmethod
    def make_sound(self):
        pass
    
    @abstractmethod
    def eat(self, food):
        pass
    
    def get_info(self):
        return f"{self.name} ({self.species}) - Age: {self.age}"
    
    @property
    def health(self):
        return self.__health
    
    @health.setter
    def health(self, value):
        if 0 <= value <= 100:
            self.__health = value

# 2. Concrete Classes (Inheritance)
class Lion(Animal):
    def make_sound(self):
        return f"{self.name} roars: RRROOAAARRR! 🦁"
    
    def eat(self, food):
        if food.lower() == "meat":
            self.health = min(self.health + 10, 100)
            return f"{self.name} enjoys the {food}!"
        return f"{self.name} refuses {food}!"

class Elephant(Animal):
    def make_sound(self):
        return f"{self.name} trumpets: PAAOOOO! 🐘"
    
    def eat(self, food):
        if food.lower() == "plants":
            self.health = min(self.health + 15, 100)
            return f"{self.name} loves eating {food}!"
        return f"{self.name} refuses {food}!"

class Monkey(Animal):
    def make_sound(self):
        return f"{self.name} screeches: OOH OOH! 🐵"
    
    def eat(self, food):
        if food.lower() in ["fruit", "banana"]:
            self.health = min(self.health + 20, 100)
            return f"{self.name} loves {food}!"
        return f"{self.name} refuses {food}!"

# 3. Using Polymorphism
zoo = [Lion("Simba", "Lion", 8), Elephant("Dumbo", "Elephant", 15), Monkey("Choco", "Monkey", 5)]

print("=" * 50)
print("🦁 ZOO ANIMALS 🐘")
print("=" * 50)

for animal in zoo:
    print(f"\n{animal.get_info()}")
    print(animal.make_sound())
    print(animal.eat("meat" if isinstance(animal, Lion) else "plants" if isinstance(animal, Elephant) else "fruit"))
    print(f"Health: {animal.health}%")
````
